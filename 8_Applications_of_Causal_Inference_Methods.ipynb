## Difference-In-Differences or Fixed-Effects Estimation for Transportation Infrastructure Evaluation

It is essential to understand the relationship between the large infrastructure spending and the level of economic activity.
More reliable evidence on them may help us better understand the impacts of infrastructure development and facilitate more rigorous evaluations of infrastructure development plans.
Using historical data on transport infrastructure construction and economic activity at the county, city, or parcel levels, there are numerous applications of difference-in-differences (or fixed effects) using spatial longitudinal data or panel data to uncover key facts.

## Relationship between highways and the
economy (Chandra & Thompson, 2000)

They examined the relationship between interstate highway construction and the level of economic activity.
Data on (i) the set of non-metropolitan counties in US that received an interstate highway after 1969, (ii) the set of non-metropolitan counties adjacent to these counties receiving a highway benefit, and (iii) the set of non-metropolitan counties that never had an interstate and are not adjacent to a county receiving an interstate as of 1994.
The first two sets of counties constitute the treatment groups for the regression analysis, and the set of counties that were neither highway nor adjacent counties and that had not received an interstate highway were used as the control group in the regressions.
It is found that highways have a differential impact across industries: certain industries grow as a result of reduced transportation costs, 
whereas others shrink as economic activity relocates.

## Example of DID: Impact of A High-Speed Railway on Residential Land Prices (Kanasugi & Ushijima, 2017)

By using the case of the high-speed railway scheduled to open in 2027 in Japan, this study examines whether the value of transport innovation is capitalized in land prices immediately after the construction plan is announced.
Adopted a hedonic approach to measure value, using balanced panel data on residential land prices from 2008 to 2015 in Japan.
Residential land prices where the time distance to the Tokyo metropolitan area reduces rose, except where the population is decreasing.
Thus, economic benefits are capitalized in land prices when demand to shorten the time distance exists.

## Site Selection Bias

Site selection bias is a significant issue in the evaluation of transport infrastructure impacts.
- One of the primary purposes of constructing transport infrastructure is to reduce transportation costs for local residents.
- Hence, transportation infrastructure is usually built in areas where people would frequently use it.
Therefore, the observed impact of the infrastructure project includes the effects due to site selection (i.e., site characteristics).
In other words, a change in the outcome (e.g., economic output) may reflect both the impacts of the infrastructure construction and the characteristics of the area where the site is located.

## Instrumental Variable Approach for Transportation Infrastructure Evaluation

It is important to develop instruments for the assignment of transportation improvements that plausibly satisfy the exclusion restriction of affecting the economic outcome of interest only through the transportation improvement.
The existing literature has followed three main instrumental variables strategies.
- The planned route instrumental variable approach, is an instrumental variables strategy which relies on planning maps and documents as a source of quasi-random variation in the observed infrastructure.
- The historical route instrumental variable approach, relies on very old transportation routes as a source of quasi-random variation in observed infrastructure.
- The inconsequential place approach, relies on choosing a sample that is inconsequential in the sense that unobservable attributes do not affect the placement of infrastructure.
(Source) Redding & Turner (2015)

## Did Highways Cause Suburbanization?
(Baum-Snow, 2007)

Assessing the extent to which the construction of new limited access highways has contributed to central-city population decline.
1950 – 1990 (highway-induced suburbanization?):
- Interstate Highway System (IHS) was constructed
- Aggregate population in central cities in the US declined by 17%, 
while population in MSAs (Metropolitan Statistical Areas) rised by 72%
- What kind of causal relationship?
Exogeneity of the interstate highway system is central to the study.
The author used planned portions of the interstate highway system as a source of exogenous variation and conducted empirical estimates
Results:
- Yes, Causal impact from IHS on suburbanization
- Additional interstate through city causes Central Cities 18% population reduction, all else equal.

<!-- slide 18 -->

## The Routes Planned in 1947 was used as instruments for estimating the impacts of interstate highways construction 1950-1990

Baum-Snow (2007)
Attaining consistent estimates of the effect of highways on suburbanization requires instrumenting for actual urban highway construction.
The routes planned in 1947 were used by the author as instruments for estimating the impacts of interstate highways construction between 1950-1990.
The validity of the 1947 plan as an instrument depends on the fact that the portion of the system in the plan was designed to facilitate trade and national defense, not to facilitate metropolitan area development.
Further, the validity requires the availability of data on control variables that influence suburbanization and are correlated with the plan.
Finally, the validity requires that planned routes be a good predictor of actual routes constructed.

## Effects of the Individual Residential Location on their Travel Behavior (Cao et al., 2010)

Most studies in the field are based on observational data, in which respondents self-select to live in, rather than being randomly assigned into, different environments.
Therefore, comparing travel behavior of people living in the environments without controlling for residential self-selection tends to produce a biased estimate of the influence of the built environment on travel behavior
This study applies propensity score matching to answer the following questions What is the size of the influences of residential location on driving?
Analyzed using Propensity Score Matching

## Propensity Score Matching: Residential Self-selection

Cao et al. (2010)
This study aims to test the impacts of residential location, which is classified based on the distance from residences to a central point, on travel behavior.
Based on the distance to the city center, the location of residences was classified into four types: urban areas, innerring suburbs, suburbs, 
and exurbs.
This study applies PSM with multiple nominal treatments, where the PS can be estimated separately for each of six pairs of control and treatment.
The dependent variable is vehicle miles driven (VMD) per day.
For each pair of locations (such as urban vs. exurban), they estimated a binary logit model using the subsample containing respondents living in the locations.

## Regression Discontinuity Design: Regulatory
Costs for House Owners (Koster et al., 2012)

Many cities apply planning policies to protect a valuable building stock.
This paper estimates the costs of within-city regulatory restrictions for house owners.
To avoid endogeneity issues due to supply restrictions, they employ a regression-discontinuity design using a World War II bombing boundary within the city of Rotterdam.
A part of the city of Rotterdam was bombed in WW II and Restrictions are more stringent in the not bombed area than in the bombed area because of the presence of historic amenities.

## Regression Discontinuity Design: Regulatory
Costs for House Owners (Koster et al., 2012)

They regress the logarithm of residential floor space price $p$ 
on a dummy $b$ indicating whether an area was bombed in 1940. 
The boundary represents an exogenous difference in policies.
To control for amenities they include a wide range of variable $x$ encompassing structural attributes, neighborhood variables, historic amenities and transaction year dummies $t$ :
It is expected that at the boundary, benefits of regulation are continuous, which is intuitive, because two houses just at the opposite side of the boundary experience the same (external) benefits.

$$\log p_s = \gamma b_s + \mathbf{z}_s'\beta + \mathbf{t}_s'\theta + \varepsilon_s$$

## Empirical Strategy (Koster et al. 2012)

Conditional on amenities and housing attributes, in the bombed area (where fewer restrictions apply) house prices are about 10% higher.
This implies regulatory costs of about 0.72 million Euro per hectare for the area under consideration.
The results suggest that house owners' benefits should be substantial to compensate for the costs of additional restrictions.
