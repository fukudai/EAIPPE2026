## Linear Regression

$Y$ : Effect (response/dependent variable) – Scalar (e.g., Ice-cream consumption)

$\bf{x}$ ’s: Causes (explanatory/independent variables) - Vector, for example
- $x_{1}$ : Temperature
- $x_{2}$ : Economic status
- $x_{3}$ : Age etc.

Possible to evaluate the marginal impact of explanatory variable $x$ on outcome $Y$ .

(Note) 
- Causal relationship cannot be identified. 
- Only correlation relationship are uncovered.

## Definition of Variables

Model: 

$$Y_i = \beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2} + \cdots + \beta_q X_{iq} + \varepsilon_i$$

- $Y_i$: Dependent Variable for observation $i$ ($i=1,2,\ldots,n$)
- $X_i$'s: Independent Variables for $i$
- $\varepsilon_i$: Error Term ($i=1,2,\ldots,n$)
- $\beta$'s: Unknown Coefficient Parameters

Number of observations = $n$

Number of parameters = $q + 1 \equiv p$


## Matrix & Vector Notation

Stacking all observations with the following matrix and vector notation:

$$\mathbf{y}=\begin{bmatrix} y_1 \\ \vdots \\ y_n \end{bmatrix},\quad \mathbf{X}=\begin{bmatrix} 1 & X_{11} & X_{12} & \cdots & X_{1q} \\ \vdots & & & \cdots & \vdots \\ 1 & X_{n1} & X_{n2} & \cdots & X_{nq} \end{bmatrix},\quad \boldsymbol{\beta}=\begin{bmatrix} \beta_0 \\ \vdots \\ \beta_q \end{bmatrix},\quad \boldsymbol{\varepsilon}=\begin{bmatrix} \varepsilon_1 \\ \vdots \\ \varepsilon_n \end{bmatrix}$$

Model:

$$\mathbf{y} = \mathbf{X}\boldsymbol{\beta} + \boldsymbol{\varepsilon}$$

**Note:** capital $Y_i$ is the dependent variable as a *random variable*; lowercase bold $\mathbf{y}=(y_1,\dots,y_n)'$ is its vector of *realized observations* (capital $X$ denotes the design matrix).


## Primary Assumptions

1. $E(\varepsilon_i)=0 \quad \forall i \quad\text{i.e.}\quad E(\boldsymbol{\varepsilon})=\mathbf{0}$
2. $E(\varepsilon_i^2)=\sigma^2 \quad \forall i$
3. $Cov(\varepsilon_i, \varepsilon_j) = E(\varepsilon_i\varepsilon_j)=0 \quad \forall i \neq j$

Assumptions 2 and 3 are combined as:
$V(\boldsymbol{\varepsilon}) = E(\boldsymbol{\varepsilon}\boldsymbol{\varepsilon}') = \sigma^2 \mathbf{I} $

Furthermore, we can impose Assumption that $\boldsymbol{\varepsilon}$ follows a multivariate normal distribution. 

Then, Assumptions 1-4 are summarized as $\boldsymbol{\varepsilon} \sim N(\mathbf{0},\ \sigma^2 \mathbf{I})$.


## Ordinary Least Squares (OLS) Method

$$\min_{\boldsymbol{\beta}} S(\boldsymbol{\beta}) = (\mathbf{y}-\mathbf{X}\boldsymbol{\beta})'(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})$$

Since the function $S(\cdot)$ is a convex function, the first-order conditions for minimization is given by:

$$\frac{\partial S(\boldsymbol{\beta})}{\partial \boldsymbol{\beta}} = \mathbf{0}$$


### OLS Estimator


$$\frac{\partial S(\boldsymbol{\beta})}{\partial \boldsymbol{\beta}} = -2\mathbf{X}'\mathbf{y} + 2\mathbf{X}'\mathbf{X}\boldsymbol{\beta} = \mathbf{0}$$

$$\Rightarrow\; \hat{\boldsymbol{\beta}} = (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\mathbf{y}$$


## Variance-Covariance Matrix of $\hat{\boldsymbol{\beta}}$

$$V(\hat{\boldsymbol{\beta}}) = E\!\left[(\hat{\boldsymbol{\beta}}-\boldsymbol{\beta})(\hat{\boldsymbol{\beta}}-\boldsymbol{\beta})'\right] = \cdots =  \sigma^2 (\mathbf{X}'\mathbf{X})^{-1}$$

where $\sigma^2$ is the variance of the error term.

## Maximum Likelihood Estimation (MLE)

In addition to Assumptions 1-3, we further assume that Assumption 4 is imposed. 

$$\boldsymbol{\varepsilon} \sim N(\mathbf{0},\ \sigma^2 \mathbf{I})$$

Then, the Joint Probability Density of error terms is:

$$f(\boldsymbol{\varepsilon}) = \frac{1}{(2\pi\sigma^2)^{n/2}} \exp\!\left(-\frac{\boldsymbol{\varepsilon}'\boldsymbol{\varepsilon}}{2\sigma^2}\right)$$

The is called Likelihood Function when it is regarded as a function of unknown parameters $\boldsymbol{\beta}$ and $\sigma$.

Since the likelihood function is hard to handle, instead its log-transformed one (Log-Likelihood Function) is utilized:
 $$\begin{aligned} L(\boldsymbol{\beta}, \sigma) &= -\frac{n}{2}\ln 2\pi - \frac{n}{2}\ln\sigma^2 - \frac{1}{2\sigma^2}\boldsymbol{\varepsilon}'\boldsymbol{\varepsilon}\\ &= -\frac{n}{2}\ln 2\pi - \frac{n}{2}\ln\sigma^2 - \frac{1}{2\sigma^2}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})'(\mathbf{y}-\mathbf{X}\boldsymbol{\beta}) \end{aligned}$$

## Maximum Likelihood Estimator (MLE)


$$\frac{\partial L(\boldsymbol{\beta}, \sigma)}{\partial \boldsymbol{\beta}} = \mathbf{0}, \quad \frac{\partial L(\boldsymbol{\beta}, \sigma)}{\partial \sigma^2} = 0$$


$$\begin{aligned} \Rightarrow &\tilde{\boldsymbol{\beta}} = (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\mathbf{y}\\ &\tilde{\sigma}^2 = \frac{1}{n}(\mathbf{y}-\mathbf{X}\tilde{\boldsymbol{\beta}})'(\mathbf{y}-\mathbf{X}\tilde{\boldsymbol{\beta}}) \end{aligned}$$


## OLS and ML Estimators

$$\begin{aligned} \hat{\boldsymbol{\beta}} &= \tilde{\boldsymbol{\beta}} = (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\mathbf{y}\\ V(\hat{\boldsymbol{\beta}}) &= V(\tilde{\boldsymbol{\beta}}) = \sigma^2(\mathbf{X}'\mathbf{X})^{-1} \end{aligned}$$

Note: Among unknown parameters, 
- OLS and ML Estimators of $\boldsymbol{\beta}$ are equivalent. 
- OLS can only estimate $\boldsymbol{\beta}$.
- ML can estimate both $\boldsymbol{\beta},\ \sigma^2$.


## Small (Finite) Sample Property of OLS Estimator

### 1. Unbiasedness (不偏性)

Unbiased Estimator:

$$E(\hat{\boldsymbol{\theta}}) = \boldsymbol{\theta}$$


OLS Estimator:

$$\begin{aligned} \hat{\boldsymbol{\beta}} &= (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\mathbf{y}\\ &= (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'(\mathbf{X}\boldsymbol{\beta}+\boldsymbol{\varepsilon})\\ &= \boldsymbol{\beta} + (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\boldsymbol{\varepsilon} \end{aligned}$$

Therefore, 
$$E(\hat{\boldsymbol{\beta}}) = \boldsymbol{\beta}$$


### 2. Best Linear Unbiasedness (最良線形不偏性)

Best Linear Unbiased Estimator (BLUE):

The unbiased estimator that exhibits the minimum variance among linear unbiased estimators

$$\hat{\boldsymbol{\theta}} = \mathbf{A}\mathbf{y} (\mathbf{y}: \text{Vector of observations})$$

$$E(\hat{\boldsymbol{\theta}}) = \boldsymbol{\theta}$$

$$\min.\ E\!\left[(\hat{\boldsymbol{\theta}}-\boldsymbol{\theta})^2\right] (\text{Mean Squared Error})$$

OLS Estimator is BLUE (from Gauss-Markov Theorem)


### 3. Efficiency (有効性)

Efficient Estimator:

An estimator that attains the minimum variance among all possible unbiased estimators

OLS Estimator:

It is an efficient estimator if $\boldsymbol{\varepsilon}$ follows a multivariate normal distribution (\because of Cramér-Rao Inequality).

## Large Sample (Asymptotic) Properties of OLS Estimator

The properties when the sample size $n$ may grow indefinitely.

### 1. Asymptotic Unbiasedness

_Asymptotically Unbiased Estimator:_

$$\lim_{n\to\infty} E(\hat{\boldsymbol{\theta}}_n) = \boldsymbol{\theta}$$


- OLS Estimator is eventually an asymptotic unbiased estimator since it is an unbiased estimator.


### 2. (Asymptotic) Consistency ([漸近]一致性)

_Consistent Estimator:_

$$\lim_{n\to\infty} E(\hat{\boldsymbol{\theta}}_n) = \boldsymbol{\theta}$$

and

$$\lim_{n\to\infty} Var(\hat{\boldsymbol{\theta}}_n) = \mathbf{0}$$

are hold, it is called a (asymptotically) consistent estimator.

The above two equations are summarized as $\operatorname*{plim}_{n\to\infty} \hat{\boldsymbol{\theta}}_n = \boldsymbol{\theta}$.

- OLS Estimator is a consistent estimator in general.


### 3. Asymptotic Normality

_Asymptotically Normal Estimator:_

- An estimator of which $\widehat{\mathbf{θ}}$ follows a multivariate normal distribution

- OLS Estimator is an asymptotically normal estimator


### 4. Asymptotic Efficiency

_Asymptotically Efficient Estimator:_

- An estimator of which the variance of the asymptotic distribution of $\widehat{\mathbf{θ}}$ (i.e., normal) is identical to the lower-bounds of Cramér-Rao

- OLS Estimator is an asymptotically efficient estimator since it is an efficient estimator


## Unbiased Estimator of $\sigma^{2}$

The residual of the OLS formula:

$$\mathbf{e} \equiv \mathbf{y} - \mathbf{X}\hat{\boldsymbol{\beta}} = \mathbf{y} - \mathbf{X}\tilde{\boldsymbol{\beta}}$$

Then we have:
$$E(\mathbf{e}'\mathbf{e}) = \sigma^2(n-p)$$

Therefore: 
$$E\!\left(\frac{\mathbf{e}'\mathbf{e}}{n-p}\right) = \sigma^2$$

Thus, the unbiased estimator of $\sigma^{2}$ is:

$$s^2 = \frac{\mathbf{e}'\mathbf{e}}{n-p}$$


## Property of ML Estimator ${\widetilde{\sigma}}^{2}$


ML Estimator is given by:  

$$\tilde{\sigma}^2 = \frac{(\mathbf{y}-\mathbf{X}\tilde{\boldsymbol{\beta}})'(\mathbf{y}-\mathbf{X}\tilde{\boldsymbol{\beta}})}{n} = \frac{\mathbf{e}'\mathbf{e}}{n}$$

Thus, ${\widetilde{\sigma}}^{2}$ is not an unbiased estimator. Rather, it is negatively biased meaning underestimation.


## Confidence Interval of $\boldsymbol{\beta}$

- A more formal indication of the precision of the estimators of the parameters.
- Reliabilities of estimated parameters can be evaluated by constructing confidence intervals.

The OLS and ML estimators exhibit: 
$$V(\hat{\boldsymbol{\beta}}) = \sigma^2(\mathbf{X}'\mathbf{X})^{-1}$$

Then we subtract the diagonal elements of the above matrix as :
$$\operatorname{diag}(\mathbf{X}'\mathbf{X})^{-1} = (q_{11}, q_{22}, \ldots, q_{pp})$$

Consequently, 
$$\hat{\beta}_j \sim N(\beta_j, \sigma^2 q_{jj}) \ \forall j = 1, \ldots, q$$
Since $\sigma$ is unknown, it is necessary to replace it with its unbiased estimator $s$ .

$$\hat{\beta}_j \sim N(\beta_j, \sigma^2 q_{jj}) \;\longrightarrow\; \frac{\hat{\beta}_j - \beta_j}{\sigma\sqrt{q_{jj}}} \sim N(0,1) \;\longrightarrow\; \frac{\hat{\beta}_j - \beta_j}{s\sqrt{q_{jj}}} \sim t_{n-p}$$

Since the prediction is always incomplete, analysts set up the degree of reliability a priori

With a probability of $(1 - \lambda)$, the following inequality is hold:

$$-t_{n-p,\,\lambda/2} \le \frac{\hat{\beta}_j - \beta_j}{s\sqrt{q_{jj}}} \le +t_{n-p,\,\lambda/2}$$
where $t_{n-p}$ is the Student's t-distribution with the degree of freedom of $n-p$.

Or equivalently,

$$\hat{\beta}_j - t_{n-p,\,\lambda/2}\, s\sqrt{q_{jj}} \;\le\; \beta_j \;\le\; \hat{\beta}_j + t_{n-p,\,\lambda/2}\, s\sqrt{q_{jj}}$$

- $1 - \lambda$: Confidence Level (The probability that the specified confidence interval covers the true value of the regression parameter)
- $\lambda$: Significance level (The most used significance levels are 5% and 1%, each of which correspond to 95% and 99% confidence levels, respectively.)

## Statistical Hypothesis Testing

- The null hypothesis is $H_0:\ \beta_j = 0 \quad \forall j$

- Also, we know the fact that: $\frac{\hat{\beta}_j - \beta_j}{s\sqrt{q_{jj}}} \sim t_{n-p}$

- Thus, if $H_0$ is true, we have: $t_j = \frac{\hat{\beta}_j}{s\sqrt{q_{jj}}} \sim t_{n-p}$

- This is called "t-value", which follows a $t$ -distribution with ($n - p$) degrees of freedom

$$t_j = \frac{\hat{\beta}_j}{s\sqrt{q_{jj}}} \sim t_{n-p} \;\longrightarrow\; -t_{n-p,\,\lambda/2} \le t_j \le +t_{n-p,\,\lambda/2}$$

If the deviation of the t-value from zero is so great as to occur by chance only 5% of the time, we shall reject the null hypothesis.

Typical example:
- e.g., $\lambda = 0.05$ (95% Confidence Level = 5% Significance Level)
$$\begin{aligned} t_{n-p,\,\lambda/2} &= 1.96 \quad (\text{if } n-p=\infty)\\ &= 2.00 \quad (\text{if } n-p=60) \end{aligned}$$

## Prediction with Linear Regression

### Settings

Model:
$$\mathbf{y} = \mathbf{X}\boldsymbol{\beta} + \boldsymbol{\varepsilon}$$

with the assumptions of 
$$E(\boldsymbol{\varepsilon})=\mathbf{0},\qquad E(\boldsymbol{\varepsilon}\boldsymbol{\varepsilon}')=\sigma^2\mathbf{I}$$

Parameter Estimation results with OLS or ML:

$$\hat{\boldsymbol{\beta}} = (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\mathbf{y}$$

Now we would like to predict the future variable of $y$ as: 

$$\hat{Y}_f = \mathbf{x}_f'\hat{\boldsymbol{\beta}}$$

with the $\mathbf{x}_f$ (explanatory variable vector in the future)

### Strict Model of Future Prediction

$$Y_f = \mathbf{x}_f'\boldsymbol{\beta} + \varepsilon_f$$


$\mathbf{x}_f$: Independent variable(s) assumed in the future

$Y_f$: Dependent variable in the future

$\varepsilon_f$: Error in the future

What is the prediction?

To obtain ${\widehat{Y}}_{f}$ , which is an estimate of $Y_{f}$

## Best Linear Unbiased Prediction (BLUP) - Desirable property for prediction

The prediction would be desirable if: 

1. Linearity $\hat{Y}_f = \mathbf{a}'\mathbf{y}$
2. Unbiasedness $E(\hat{Y}_f - Y_f) = 0$
3. The Best Prediction in the sense that "Mean least square error is minimized (that is equivalent to the minimization of the variance of prediction error) as": $\min.\ E\!\left[(\hat{Y}_f - Y_f)^2\right]$
### Assumption

$$Y_f = \mathbf{x}_f'\boldsymbol{\beta} + \varepsilon_f$$

This is an analogy to $E(\boldsymbol{\varepsilon})=\mathbf{0},\qquad E(\boldsymbol{\varepsilon}\boldsymbol{\varepsilon}')=\sigma^2\mathbf{I}$.

Then finally we have:
$$E(\varepsilon_f) = 0$$
$$E(\varepsilon_f^2) = \sigma^2$$
$$E(\varepsilon_f \boldsymbol{\varepsilon}) = \mathbf{0}$$
### BLUP

BLUP is then given by:
$$\hat{Y}_f = \mathbf{x}_f'\hat{\boldsymbol{\beta}}$$
together with the variance of the prediction error (minimum variance) as:
$$\hat{\sigma}_f^2 = \sigma^2\!\left[1 + \mathbf{x}_f'(\mathbf{X}'\mathbf{X})^{-1}\mathbf{x}_f\right]$$

Thus, it is possible to evaluate the error or the reliability of the future prediction by regression.
