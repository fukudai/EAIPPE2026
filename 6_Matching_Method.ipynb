## What is the “Matching” Method?

The basic idea of multiple regression for causal analysis is that 
we measure the policy impacts while controlling external factors.
There is another way of controlling external factors that is more straightforward. It is to compare those with similar external factors.
The matching method is the way of appropriately collecting those who have similar external factors.

## Introducing “Experimental Approach” for Causal Analysis

One of the most remarkable changes in empirical analysis studies is to promptly utilize experimental approach in the real field.
Randomized Controlled Trial (RCT): Construct the control group by lotteries
- Effects of constructing irrigation equipment on economic development (Banerjee & Duflo 2011)
- Effects of institutional design of small-scale financing on repayment rate (Ashraf et al. 2006)
By coinciding (all) external conditions, it tries to unveil the causal relationship which is considered to be hard to examine.
- External Validity is the desirable property that the conclusion derived from the experimental data analysis can be applied to environment other than the situation of the experiments.
- To ensure the external validity to some extent, it is necessary to increase sample size, anyway.
- However, that is actually costly.

## Application of RCT: Health

One important issue is whether medicine and healthcare should be charged for and, if so, what they should cost.
A field experiment by Kremer and co-author investigated how the demand for deworming pills for parasitic infections (寄生虫駆虫のための投薬行為) was affected by price.
They found that 75% of parents gave their children these pills when the medicine was free, compared to 18 per cent when they cost less than a US dollar, which is still heavily subsidized.
Subsequently, many similar experiments have found the same thing: poor people are extremely price-sensitive regarding investments in preventive healthcare.

## Application of RCT: Health

Low service quality is another explanation why poor families invest so little in preventive measures. One example is that staff at the health centers that are responsible for vaccinations are often absent from work.
Banerjee, Duflo & their colleagues investigated whether mobile vaccination clinics – where the care staff were always on site – could fix this problem. Vaccination rates tripled in the villages that were randomly selected to have access to these clinics, at 18 % compared to 6 %.
This increased further, to 39 %, if families received a bag of lentils as a bonus when they vaccinated their children. Because the mobile clinic had a high level of fixed costs, the total cost per vaccination actually halved, despite the additional expense of the lentils.

## Basic Idea on Matching

To investigate a policy effect, we firstly focus on those who are affected by that policy (Person A).
Next, among the people who were not intervened by the policy, we look for those who are similar with Person A. We call him/her Person B. Then we regard Person B (who is similar with Person A) to be the counterfactual of Person A who might have been realized if Person A was not intervened.
The policy impact for Person A can be evaluated as the difference in the outcomes between Person A (that is realized) and Person B (that is the counterfactual of Person A).

<!-- slide 10 -->

## Average Treatment Effect (ATE)

The expression "treatment effect" refers to the causal effect of a given treatment or intervention (e.g., the administering of a drug) on an outcome variable of interest (e.g., the health of the patient).
Define two potential outcomes: 
 $Y_{0i}$ is the value of the outcome variable for individual $i$ if they are not treated, 
 $Y_{1i}$ is the value of the outcome variable for individual $i$ if they are treated. 
We also call them “Potential Outcome Variables”.
The causal effect for $i$ is the difference of the two outcomes: $Y_{1i} - Y_{0i}$
Average Treatment Effect in the population: $E\left\lbrack Y_{1i} - Y_{0i} \right\rbrack$
ATE can be estimated with the sampled dataset as its mean: $\frac{1}{N}\sum_{i = 1}^{N}\left( Y_{1i} - Y_{0i} \right)$

<!-- slide 11 -->

## Average Treatment effect on the Treated (ATT)

We are often interested in the impact of the policy especially for those who are affected by the policy.
$Y_{1i}\$ : Outcome if $i$ is intervened by the policy, 
 $Y_{0i}$ : Outcome if $i$ is NOT intervened by the policy
The causal effect for $i$ is the difference of the two outcomes: $Y_{1i} - Y_{0i}$
Average Treatment Effect on the Treated: $E\left\lbrack Y_{1i} - Y_{0i}|\ X_{i} = 1 \right\rbrack$ 
( $X_{i} = 1$ if treated and 0 for not treated)
ATT can be estimated with the sampled dataset as its mean: $\frac{1}{\sum X_{i}}\ \sum X_{i} \times \left( Y_{1i} - Y_{0i} \right)$
Actually, either $Y_{1i}$ or $Y_{0i}$ can be observed for each individual $i$ , so we call them potential outcome variables. We thus cannot understand the causal impact completely

## Matching Method

The most serious issue on ATE (ATT): Actually, either $Y_{1i}$ or $Y_{0i}$ can be observed for each individual $i$ .
Alternatively, we look for the two (or more) similar individuals: one from the control group and another from the treatment group. 
Then we compare the outcome differences between these two individuals.
How could we evaluate the similarity between the two individuals?
- Matching with the covariates 
(e.g., Breakfast - Test Score: we find two households from each group with the same (or similar) income)
- Not necessarily restrict to a single covariates but we may consider multiple covariates.
- The similarity (or the distance of person $i$ and $j$ ) can be measured by the (Mahalanobis’) Distance. 
For the case of two covariates, the normal distance is derived as: 

 $\sqrt{\left( C_{1i} - C_{1j} \right)^{2} + \left( C_{2i} - C_{2j} \right)^{2}}$ 

that can accommodate the difference in the variance across two attributes.

<!-- slide 13 -->

## Various Matching Methods

Complete Matching: matched if and only if the distance between the covariates is zero (In many cases, however, it is hard to conduct)
Nearest Neighbor Matching: two individuals with the shortest distance
Radius-based Matching: individuals within a predetermined radius of the distance.
Kernel-based Matching: = Radius-based Matching weighted by Kernel Function
The robustness analysis is needed by trying various matching methods.

<!-- slide 14 -->

## Two Important Assumptions for Appropriate Matching Method

(1) Conditional Independent Assumption 
Potential outcome variable and policy variable are independent if conditioned by external conditions.
- Basically, the same with the regression analysis that the policy variable and the error terms are independent if it is conditioned by the external variable $C_{i}$
(2) Common Support Assumption
For each covariate $C$ , there must be both who are affected by the policy and who are not affected by the policy. It can be formulated as:
$$0 < \text{Prob}\left( X_{i} = 1\ \middle| {\ C}_{i} \right) < 1$$
- For the individuals in the control group, if there are no individuals with the same covariates 
in the treatment group, it is not possible to find the individual similar with him/her. 
Thus, matching method cannot evaluate the causal impacts.

## Propensity Score Matching (PSM)

Most commonly-used practical approach
It becomes difficult to find individuals with the same combinations of multiple covariates between control and treatment groups.
- There are no two individuals with the same gender, the same age, the same educational background, the same income, the same family composition etc.…
PSM = Matching individuals with the probability that each individual is treated.
- The individuals with the same (similar) treated probabilities can be regarded as “(almost) the same person”, even though their attributes (covariates) are different.
- The treated probability can be models as a function of observed attributes of individual.
- Easy to practice because it is a single-value (i.e., treated probability) based matching.

## Propensity Score Matching - How it Works?

The treatment group and the control group are identified.
The outcome variable and its relevant predictor variables (that are available in the data) are specified.
The probability of being in the treatment group is estimated for each subject in the sample, given their observable characteristics. Binary probit / logit models are widely used.
Treatment groups and control groups are matched based upon their likelihood (probabilities) to receive the treatment.
The treatment effect is estimated as the difference between the treatment groups and their statistically-matched control groups.

## Example of PSM – Impact of Education on Earnings

Research Question: How much is the income raised by university graduation?
- University graduates tends to have originally higher abilities. It consequently leads to the increase in their income.
- If it is the case, income raising may not be the effect of schooling.
Identify those who have the similar “probabilities” of going to universities and match them from both control and treatment groups. Then, compare their incomes to evaluate effects of university schooling.
Example Dataset (10_1_income.csv) with 4371 individuals from Tanaka (2015).
- $\text{𝑐𝑜𝑔𝑟𝑎𝑑}$ : 1 if he/she graduated from universities and 0 for not.
- $\text{𝑝𝑎}\text{𝑐𝑜𝑔𝑟𝑎𝑑}$ : 1 if his/her father graduated from universities and 0 for not.
- $\text{𝑠𝑖𝑏𝑠}$ : number of brothers/sisters
- $\text{𝑙𝑖𝑛𝑐𝑜𝑚𝑒}$ : natural log of annual income
- We control individuals by $\text{𝑝𝑎𝑐𝑜𝑔𝑟𝑎𝑑}$ and $\text{𝑠𝑖𝑏𝑠}$

## Example of PSM – Impact of Education on Earnings

Compute the Propensity Score with the “probit regression”:
$c\text{𝑜𝑔𝑟𝑎}d_{i} = \left\{ \begin{matrix}
1,\ \ \text{if}\ \text{𝑐𝑜𝑔𝑟𝑎}d_{\text{𝑙𝑎𝑡𝑒𝑛𝑡},\ i} \geq 0 \\
\& 0,\text{if}\ \text{𝑐𝑜𝑔𝑟𝑎}d_{\text{𝑙𝑎𝑡𝑒𝑛𝑡},\ i} < 0 \\
\end{matrix} \right.\$ where, $\text{𝑐𝑜𝑔𝑟𝑎}d_{\text{𝑙𝑎𝑡𝑒𝑛𝑡},\ i} = \beta_{0} + \beta_{1}\text{𝑝𝑎𝑐𝑜𝑔𝑟𝑎}d_{i} + \beta_{2}\text{𝑠𝑖𝑏}s_{i} + u_{i}$
Note: Probit model if $u_{i}\$ follows normal distribution and Logit model if $u_{i}\$ follows Gumbel distribution
Maximum likelihood result (probit):
$$\widehat{\text{𝑐𝑜𝑔𝑟𝑎}d_{\text{𝑙𝑎𝑡𝑒𝑛𝑡},\ i}} = - 0.458 + 0.3665\text{𝑝𝑎𝑐𝑜𝑔𝑟𝑎}d_{i} - 0.125\text{𝑠𝑖𝑏}s_{i}$$
# Data Read
income10 <- read.table("data/10_1_income.csv", header = TRUE, sep = ",")
# Probit regression for computing the probability of
# being in the treatment group
PScoreModel <- glm(cograd ~ pacograd + sibs,
family= binomial(link = "probit"),
data = income10)
summary(PScoreModel)
# Compute Propensity Score
PScores <- PScoreModel$fitted.values

## Example of PSM – Impact of Education on Earnings

Match individuals with the computed propensity score with the following criteria:
- One-to-one matching
- If the probability difference exceeds 10% (=0.1), it is not matched.
- If, for example, one treated observation matches more than one control observation, the randomly chosen one control observation is matched.
# Install the package "Matching"
# install.packages("Matching")
library(Matching)
# Conduct PSM
PSM0.1 <- Match(Y = income10$lincome, # Outcome
Tr = (income10$cograd==1), # Treatment
X = PScores, # Propensity Score
M = 1, # one to one matching
caliper = 0.1, # If PS exceeds, it is not matched
ties = FALSE, # Ties will be randomly broken
replace = FALSE # sampling without replacement
)

## Summary

“Matching” is the method like that:
- Find the individual similar with those are affected by the policy 
but are actually unaffected.
- Then compare their outcomes to evaluate the causal impacts.
“Propensity Score Matching” is the most widely used in practice because it is more feasible to compare individuals with the single covariate, that is, the probability of being treated.
