## Slide 3

Endogeneity in Explanatory Variables
$$Y = \beta_{0} + \beta_{1}X_{1} + \beta_{2}X_{2} + \cdots + \beta_{k}X_{k} + U$$
The most important condition for OLS estimator to be unbiased:
 $E\left\lbrack U \middle| X_{1},X_{2},\ldots,\ X_{k} \right\rbrack = 0,$
(all explanatory variables are “mean independent” from error term)
The weaker assumption that “Error and explanatory variables are uncorrelated” ( $E\left\lbrack U \right\rbrack = 0,\ \text{𝐶𝑜𝑣}\left\lbrack X_{j},U \right\rbrack = 0,\ j = 1,\ldots,k$ ) leads to the consistency of OLS estimator as asymptotic properties ( $\text{plim}\widehat{\mathbf{β}} = {\mathbf{β}}$ )

<!-- slide 3 -->

## Slide 4

Endogeneity in Explanatory Variables
In real observational data, however, policy variables and errors would be correlated.
If the explanatory variable is correlated with error, it is called “endogenous variable.” Then, there is “endogeneity.”
[If not, it is called “exogenous variable.”]
- Endogenous: $\text{𝐶𝑜𝑣}(X,U) \neq 0$
- Exogenous: $\text{𝐶𝑜𝑣}(X,U) = 0$ [So far, we assumed $E(U|X) = 0$ ]
If there is endogeneity, the OLS estimator does NOT exhibit consistency and thus the causal analysis is NOT reliable.

<!-- slide 4 -->

## Instrumental Variable (IV) Method

Approach to obtain consistent estimator with endogeneity in explanatory variables.
- Available for various endogeneity bias due to “omitted variables”, “measurement errors”, “simultaneity” etc.
Instrumental variables (shortly “instruments”) $Z$ must be:
- Uncorrelated with error ( $\text{𝐶𝑜𝑣}(Z,U) = 0$ ) and,
- Correlated with exogenous variable ( $\text{𝐶𝑜𝑣}\left( Z,X \right) \neq 0$ ).

## IV Estimator (Simple Linear Regression)

Model: $Y = \beta_{0} + \beta_{1}X + U$ and $\text{𝐶𝑜𝑣}(X,U) \neq 0$
For the instruments $Z$ , $\text{𝐶𝑜𝑣}(Z,U) = 0$ and $\text{𝐶𝑜𝑣}\left( Z,X \right) \neq 0$
By taking expectations of the model with the assumption $E\lbrack U\rbrack = 0$ , we have: $E\left\lbrack Y \right\rbrack = \beta_{0} + \beta_{1}E\lbrack X\rbrack$
By taking differences, we have: $Y - E\left\lbrack Y \right\rbrack = \beta_{1}\left( X - E\left\lbrack X \right\rbrack \right) + U$
By multiplying $(Z - E\lbrack Z\rbrack)$ and then taking expectations:
 $E\left\lbrack \left( Z - E\left\lbrack Z \right\rbrack \right)\left( Y - E\left\lbrack Y \right\rbrack \right) \right\rbrack = \beta_{1}E\left\lbrack \left( Z - E\left\lbrack Z \right\rbrack \right)\left( X - E\left\lbrack X \right\rbrack \right) \right\rbrack + E\left\lbrack \left( Z - E\left\lbrack Z \right\rbrack \right)U \right\rbrack$ 
 $\Longleftrightarrow \text{𝐶𝑜𝑣}\left\lbrack Z,Y \right\rbrack = \beta_{1}\text{𝐶𝑜𝑣}\left\lbrack Z,X \right\rbrack \Longleftrightarrow \beta_{1} = \frac{\text{𝐶𝑜𝑣}\left\lbrack Z,Y \right\rbrack}{\text{𝐶𝑜𝑣}\left\lbrack Z,X \right\rbrack}$ $\Longrightarrow \ {\widehat{\beta}}_{1} = \frac{\frac{1}{n}\sum_{i = 1}^{n}{\left( z_{i} - \overline{z} \right)\left( y_{i} - \overline{y} \right)}}{\frac{1}{n}\sum_{i = 1}^{n}{\left( z_{i} - \overline{z} \right)\left( x_{i} - \overline{x} \right)}\ }$

## Statistical Test for IV Method

Homoscedastic assumption of error term: $V\lbrack U|Z\rbrack = E\lbrack U^{2}|Z\rbrack = s^{2}$
If this is true, the variance of the IV estimator $\beta_{1}$ and the standard error are given by:
$$V\left\lbrack {\widehat{\beta}}_{1} \right\rbrack = \frac{s^{2}}{\left\lbrack \sum_{i = 1}^{n}\left( X_{i} - E\left\lbrack X \right\rbrack \right)^{2} \right\rbrack\left( \rho_{X,Z} \right)^{2}} \Longrightarrow s.e.\ ({\widehat{\beta}}_{1}) = \left( \frac{{\widehat{s}}^{2}}{\left\lbrack \sum_{i = 1}^{n}\left( X_{i} - E\left\lbrack X \right\rbrack \right)^{2} \right\rbrack R_{X,Z}^{2}} \right)^{\frac{1}{2}}$$
- $\rho_{X,Z}$ : Correlation of coefficients of $X$ and $Z$
- $R_{X,Z}^{2}$ : Determinants of coefficients squared for the simple regression $X = \delta_{0} + \delta_{1}Z + V$
- ${\widehat{s}}^{2} = \frac{1}{n - 2}\sum_{i = 1}^{n}{\widehat{u}}_{i}^{2}$ with the estimates of error variance from IV that ${\widehat{u}}_{i} = y_{i} - {\widehat{\beta}}_{0} -$ ${\widehat{\beta}}_{1}x_{i}$
The S.E. for the IV is different from the one for OLS only in that there is $R_{X,Z}^{2}$ or not.
Since $R_{X,Z}^{2} < 1$ , the S.E. for IV is always larger than those for OLS.
This implies that “The better Instruments” is the one which is strongly correlated with exogenous variable.

<!-- slide 10 -->

## How to find Better Instruments?

$\text{𝐶𝑜𝑣}\left( Z,X \right) \neq 0$ ← Can be (easily) investigated from observed data
$\text{𝐶𝑜𝑣}(Z,U) = 0$ ← Hard to verify from data, but mostly based on existing theory, common sense and persuasion by the analysts.
What if “wrong” instruments (i.e., $\text{𝐶𝑜𝑣}\left( Z,U \right) \neq 0$ ) are used?
- We now compare the asymptotic biases
  - IV: $\text{plim}{\widehat{\beta}}_{1} = \beta_{1} + \frac{\rho_{Z,U}}{\rho_{Z,X}} \times \frac{\text{𝑠𝑑}(U)}{\text{𝑠𝑑}(X)}$ OLS: $\text{plim}{\widehat{\beta}}_{1} = \beta_{1} + \rho_{X,U} \times \frac{\text{𝑠𝑑}(U)}{\text{𝑠𝑑}(X)}$
- If the correlation of $Z$ and $X$ is very weak, [or the correlation of $Z$ and $U$ is not weak] the (absolute) value of bias tends to be very large (weak instruments).
- In such cases, the use of the simple OLS might be better than the use of IV.

## 2SLS for Multiple Instruments Case

How we should if some instruments are available, say $Z_{1}$ and $Z_{2}$ ?
The solely use of either $Z_{1}$ or $Z_{2}$ is not efficient.
Instead, we estimate the first step (OLS) regression:
 $X = \delta_{0} + \delta_{1}Z_{1} + \delta_{2}Z_{2} + \delta_{3}C + V$
We obtain the fitted endogenous variable:
 ${\widehat{X}}_{i} = {\widehat{\delta}}_{0} + {\widehat{\delta}}_{1}Z_{1i} + {\widehat{\delta}}_{2}Z_{2i} + {\widehat{\delta}}_{3}C_{i}\text{~~~}i = 1,\ \ldots,n$
Then we perform the second step (OLS) regression:
 $Y_{i} = \beta_{0} + \beta_{1}{\widehat{X}}_{i} + \beta_{2}C_{i} + U_{i}\text{~~~~}i = 1,\ \ldots,n$

## R 実習コード（Class4_IV.R）

対応するRスクリプトをIRkernelで実行できます．データの相対パスは適宜調整して下さい．

```r
# Data Read
income8 <-read.table("8_income.csv", header = TRUE, sep= ",")

# The Simple OLS result
reg1 <- lm(lincome~ yeduc, data = income8)
summary(reg1)

# Check the appropriateness of instrument
reg2 <- lm(yeduc ~ payeduc, data = income8)
summary(reg2)

# Read the package
install.packages("AER")
require(AER)
# IV result
ivreg1 <- ivreg(lincome ~ yeduc | payeduc, data = income8)
summary(ivreg1)

# See what happens if inappropriate instrument is used.
# IV result
ivreg2 <- ivreg(lincome ~ yeduc | mbirth, data = income8)
summary(ivreg2)

# Regress the fitted yeduc to lincome
delta0hat <- reg2$coefficients[1]
delta1hat <- reg2$coefficients[2]
income8 <- transform(income8, 
                     yeduchat = delta0hat + delta1hat * payeduc)
reg3 <- lm(lincome ~ yeduchat, data = income8)
summary(reg3)

# Multiple IV case
ivreg3 <- ivreg(lincome ~ yeduc + exper + exper2 | 
                  exper + exper2 + payeduc + sibs, data=income8)
summary(ivreg3)

#Compare the above IV estimate of yeduc (0.0699093) 
#with the below corresponding OLS estimate (0.0842350)
reg4 <- lm(lincome ~ yeduc + exper + exper2, data=income8)
summary(reg4)

```
