2. Derive the **ordinary least squares (OLS)** estimator and interpret it as an orthogonal projection.

**Why this matters for project / policy evaluation.** Regression is the workhorse for measuring how an outcome (land price, travel demand, household income, …) responds to explanatory factors. **Notation convention.** Capital $Y$ = random variable; lowercase $y$ = its realized (observed) value; bold lowercase $\mathbf{y}$ = vector; bold capital $\mathbf{X}$ = matrix; prime $'$ = transpose; hat $\hat{\cdot}$ = estimator.

## 1. What Is Linear Regression?

$Y$ : effect (response / dependent variable) — a scalar (e.g., ice-cream consumption, land price, traffic volume).

$\mathbf{x}$ : causes (explanatory / independent variables) — a vector, for example
- $x_{1}$ : temperature
- $x_{2}$ : economic status
- $x_{3}$ : age, etc.

Linear regression describes how the **conditional mean** of $Y$ moves with the explanatory variables:
$$E[Y \mid \mathbf{x}] = \beta_0 + \beta_1 x_1 + \cdots + \beta_q x_q .$$
The coefficient $\beta_j = \partial E[Y\mid\mathbf{x}]/\partial x_j$ is the **marginal effect** of $x_j$, holding the other regressors fixed ("ceteris paribus"). The gap between the realized $Y$ and its conditional mean is the **error term** $\varepsilon$, which absorbs every other influence on $Y$ not captured by the regressors.

**(Caution) Correlation vs. causation.**
- A fitted regression uncovers *correlation / conditional association* only.
- A **causal** reading of $\beta_j$ further requires that $x_j$ be uncorrelated with $\varepsilon$ (exogeneity). In observational data this usually fails because of omitted variables, simultaneity, or measurement error.
- Recovering causal effects is precisely the agenda of the following lectures (causal effect analysis, instrumental variables, panel data, matching, quasi-experiments).

## 2. The Regression Model

**Data-generating process for observation $i$:**
$$Y_i = \beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2} + \cdots + \beta_q X_{iq} + \varepsilon_i, \qquad i = 1,2,\ldots,n.$$

- $Y_i$ : dependent variable for observation $i$
- $X_{ij}$ : $j$-th independent variable for $i$
- $\varepsilon_i$ : error (disturbance) term
- $\beta_0,\ldots,\beta_q$ : unknown coefficient parameters

**Key points.**
- The model is **linear in the parameters** $\beta_j$, *not* necessarily in the variables. Transformations such as $x^2$, $\ln x$, or interactions $x_1 x_2$ are allowed, because they enter linearly through their coefficients.
- $\beta_0$ is the **intercept** — the conditional mean of $Y$ when all regressors equal zero.
- Number of observations $= n$.
- Number of parameters $= q + 1 \equiv p$.

## 3. Matrix & Vector Notation

Stacking all $n$ observations:

$$\mathbf{y}=\begin{bmatrix} y_1 \\ \vdots \\ y_n \end{bmatrix},\quad \mathbf{X}=\begin{bmatrix} 1 & X_{11} & X_{12} & \cdots & X_{1q} \\ \vdots & & & \cdots & \vdots \\ 1 & X_{n1} & X_{n2} & \cdots & X_{nq} \end{bmatrix},\quad \boldsymbol{\beta}=\begin{bmatrix} \beta_0 \\ \vdots \\ \beta_q \end{bmatrix},\quad \boldsymbol{\varepsilon}=\begin{bmatrix} \varepsilon_1 \\ \vdots \\ \varepsilon_n \end{bmatrix}$$

Model:
$$\mathbf{y} = \mathbf{X}\boldsymbol{\beta} + \boldsymbol{\varepsilon}$$

**Dimensions.** $\mathbf{y}$ is $n\times 1$, $\mathbf{X}$ is $n\times p$, $\boldsymbol{\beta}$ is $p\times 1$, $\boldsymbol{\varepsilon}$ is $n\times 1$ (with $p=q+1$).

**Identification (full-rank condition).** We require $\operatorname{rank}(\mathbf{X})=p$ — the columns of $\mathbf{X}$ are linearly independent (which needs $n\ge p$). This makes $\mathbf{X}'\mathbf{X}$ invertible and the coefficients uniquely determined. **Perfect multicollinearity** among regressors violates it.

**Note:** capital $Y_i$ is the dependent variable as a *random variable*; lowercase bold $\mathbf{y}=(y_1,\dots,y_n)'$ is its vector of *realized observations* (capital $\mathbf{X}$ denotes the design matrix).

## 4. Classical (Gauss–Markov) Assumptions

**(A0) Full rank.** $\operatorname{rank}(\mathbf{X})=p$ — no perfect multicollinearity.

**(A1) Zero-mean error / exogeneity.** $E(\varepsilon_i)=0\ \forall i$, i.e. $E(\boldsymbol{\varepsilon})=\mathbf{0}$. The stronger *strict exogeneity* $E(\boldsymbol{\varepsilon}\mid\mathbf{X})=\mathbf{0}$ is what makes OLS unbiased and lets $\beta_j$ be read causally.

**(A2) Homoskedasticity.** $E(\varepsilon_i^2)=\sigma^2\ \forall i$ — Constant error variance across observations.

**(A3) No autocorrelation.** $Cov(\varepsilon_i,\varepsilon_j)=E(\varepsilon_i\varepsilon_j)=0\ \forall i\neq j$.

Assumptions 2 and 3 combine into the **spherical-error** condition:
$$V(\boldsymbol{\varepsilon}) = E(\boldsymbol{\varepsilon}\boldsymbol{\varepsilon}') = \sigma^2 \mathbf{I}.$$

**(A4) Normality (optional).** $\boldsymbol{\varepsilon} \sim N(\mathbf{0},\ \sigma^2 \mathbf{I})$. Required for *exact* finite-sample $t$- and $F$-distributions and for MLE; **not** needed for OLS to be the best linear unbiased estimator.

**What happens if an assumption fails (preview).**

| Violated | Consequence for OLS | Typical solutions (optional) |
|---|---|---|
| A2 Homoskedasticity | still unbiased; usual SEs invalid | robust (White) SEs, WLS/GLS |
| A3 No autocorrelation | still unbiased; usual SEs invalid | HAC SEs, GLS |
| A1 Exogeneity ($Cov(x,\varepsilon)\neq0$) | **biased and inconsistent** | IV, panel, matching, … (Lectures 3–8) |

## 5. Ordinary Least Squares (OLS)

OLS chooses $\boldsymbol{\beta}$ to minimize the **sum of squared residuals**:
$$\min_{\boldsymbol{\beta}} S(\boldsymbol{\beta}) 
= \sum_{i=1}^n \left(y_i - \beta_0 - \beta_1 X_{i1} - \beta_2 X_{i2} - \cdots - \beta_q X_{iq}\right)^2
= (\mathbf{y}-\mathbf{X}\boldsymbol{\beta})'(\mathbf{y}-\mathbf{X}\boldsymbol{\beta}).$$

Expanding the quadratic form,
$$S(\boldsymbol{\beta}) = \mathbf{y}'\mathbf{y} - 2\boldsymbol{\beta}'\mathbf{X}'\mathbf{y} + \boldsymbol{\beta}'\mathbf{X}'\mathbf{X}\boldsymbol{\beta}.$$

Because $\mathbf{X}'\mathbf{X}$ is positive semidefinite and $S(\cdot)$ is a **convex** function, so the first-order condition characterizes the global minimum:
$$\frac{\partial S(\boldsymbol{\beta})}{\partial \boldsymbol{\beta}} = \mathbf{0}.$$

### OLS Estimator

Differentiating and setting the gradient to zero gives the **normal equations**:
$$\frac{\partial S(\boldsymbol{\beta})}{\partial \boldsymbol{\beta}} = -2\mathbf{X}'\mathbf{y} + 2\mathbf{X}'\mathbf{X}\boldsymbol{\beta} = \mathbf{0} \;\Longrightarrow\; \mathbf{X}'\mathbf{X}\hat{\boldsymbol{\beta}} = \mathbf{X}'\mathbf{y}.$$

Under (A0), $\mathbf{X}'\mathbf{X}$ is invertible, hence
$$\boxed{\;\hat{\boldsymbol{\beta}} = (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\mathbf{y}\;}$$

**Geometric interpretation.** 
  
- The fitted values are $\hat{\mathbf{y}} = \mathbf{X}\hat{\boldsymbol{\beta}} = \mathbf{P}\mathbf{y}$ with the **projection ("hat") matrix** $\mathbf{P}=\mathbf{X}(\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'$. 

- The residual vector is $\mathbf{e} = \mathbf{y}-\hat{\mathbf{y}} = \mathbf{M}\mathbf{y}$ with $\mathbf{M}=\mathbf{I}-\mathbf{P}$. 

- OLS **orthogonally projects** $\mathbf{y}$ onto the column space of $\mathbf{X}$, so $\mathbf{X}'\mathbf{e}=\mathbf{0}$: residuals are orthogonal to every regressor. 

- Both $\mathbf{P}$ and $\mathbf{M}$ are symmetric and idempotent.

## 6. Variance–Covariance Matrix of $\hat{\boldsymbol{\beta}}$

Substitute the model $\mathbf{y}=\mathbf{X}\boldsymbol{\beta}+\boldsymbol{\varepsilon}$ into the estimator:
$$\hat{\boldsymbol{\beta}} = (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\mathbf{y} = \boldsymbol{\beta} + (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\boldsymbol{\varepsilon}.$$

Then, conditional on $\mathbf{X}$ and using $V(\boldsymbol{\varepsilon})=\sigma^2\mathbf{I}$,
$$V(\hat{\boldsymbol{\beta}}) = E\!\left[(\hat{\boldsymbol{\beta}}-\boldsymbol{\beta})(\hat{\boldsymbol{\beta}}-\boldsymbol{\beta})'\right] = (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\,(\sigma^2\mathbf{I})\,\mathbf{X}(\mathbf{X}'\mathbf{X})^{-1} = \sigma^2 (\mathbf{X}'\mathbf{X})^{-1}.$$

The square roots of the diagonal entries are the **standard errors** of the coefficients (after replacing the unknown $\sigma^2$ by its estimate $s^2$, introduced later). They quantify the sampling precision of each $\hat{\beta}_j$.

## 7. Maximum Likelihood Estimation (MLE)

Now we add the normality assumption (A4): $\boldsymbol{\varepsilon} \sim N(\mathbf{0},\ \sigma^2 \mathbf{I})$.

The joint probability density of the errors is
$$f(\boldsymbol{\varepsilon}) = \frac{1}{(2\pi\sigma^2)^{n/2}} \exp\!\left(-\frac{\boldsymbol{\varepsilon}'\boldsymbol{\varepsilon}}{2\sigma^2}\right).$$

which is regarded as a function of the unknown parameters $(\boldsymbol{\beta},\sigma^2)$ given the data, this is the **likelihood**. 

Because products are awkward, we maximize its logarithm, the **log-likelihood** (substituting $\boldsymbol{\varepsilon}=\mathbf{y}-\mathbf{X}\boldsymbol{\beta}$):
$$\begin{aligned} L(\boldsymbol{\beta}, \sigma^2) &= -\frac{n}{2}\ln 2\pi - \frac{n}{2}\ln\sigma^2 - \frac{1}{2\sigma^2}\boldsymbol{\varepsilon}'\boldsymbol{\varepsilon}\\ &= -\frac{n}{2}\ln 2\pi - \frac{n}{2}\ln\sigma^2 - \frac{1}{2\sigma^2}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})'(\mathbf{y}-\mathbf{X}\boldsymbol{\beta}). \end{aligned}$$

**Intuition.** Maximizing $L$ over $\boldsymbol{\beta}$ means minimizing $(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})'(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})$ — exactly the OLS objective. So under normality the ML and OLS estimators of $\boldsymbol{\beta}$ coincide.

### Maximum Likelihood Estimator

The first-order conditions
$$\frac{\partial L}{\partial \boldsymbol{\beta}} = \mathbf{0}, \qquad \frac{\partial L}{\partial \sigma^2} = 0$$
yield
$$\Rightarrow\quad \tilde{\boldsymbol{\beta}} = (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\mathbf{y}, \qquad \tilde{\sigma}^2 = \frac{1}{n}(\mathbf{y}-\mathbf{X}\tilde{\boldsymbol{\beta}})'(\mathbf{y}-\mathbf{X}\tilde{\boldsymbol{\beta}}).$$

The slope/intercept estimator $\tilde{\boldsymbol{\beta}}$ is identical to the OLS $\hat{\boldsymbol{\beta}}$; the variance estimator $\tilde{\sigma}^2$ divides by $n$ (not $n-p$), which we revisit below.

### OLS and ML Estimators

$$\hat{\boldsymbol{\beta}} = \tilde{\boldsymbol{\beta}} = (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\mathbf{y}, \qquad V(\hat{\boldsymbol{\beta}}) = V(\tilde{\boldsymbol{\beta}}) = \sigma^2(\mathbf{X}'\mathbf{X})^{-1}.$$

**Note.**
- The OLS and ML estimators of $\boldsymbol{\beta}$ are **equivalent** (under normality).
- OLS estimates only $\boldsymbol{\beta}$; the error variance is estimated separately.
- ML estimates **both** $\boldsymbol{\beta}$ and $\sigma^2$ jointly.

## 8. Finite (Small) Sample Properties of OLS

### 1. Unbiasedness
An estimator is **unbiased** if $E(\hat{\boldsymbol{\theta}}) = \boldsymbol{\theta}$. 

From $\hat{\boldsymbol{\beta}} = \boldsymbol{\beta} + (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\boldsymbol{\varepsilon}$ and exogeneity $E(\boldsymbol{\varepsilon}\mid\mathbf{X})=\mathbf{0}$, we have:
$$E(\hat{\boldsymbol{\beta}}\mid\mathbf{X}) = \boldsymbol{\beta} + (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\,E(\boldsymbol{\varepsilon}\mid\mathbf{X}) = \boldsymbol{\beta}.$$

### 2. Best Linear Unbiasedness (Gauss–Markov theorem)
Among all **linear unbiased** estimators $\hat{\boldsymbol{\theta}} = \mathbf{A}\mathbf{y}$, OLS has the smallest variance — it is **BLUE** (Best Linear Unbiased Estimator).

*Proof sketch.* Write any competing linear unbiased estimator as $\tilde{\boldsymbol{\beta}} = [(\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}' + \mathbf{C}]\mathbf{y}$. Unbiasedness for all $\boldsymbol{\beta}$ forces $\mathbf{C}\mathbf{X}=\mathbf{0}$. A direct computation then gives
$$V(\tilde{\boldsymbol{\beta}}) = \sigma^2(\mathbf{X}'\mathbf{X})^{-1} + \sigma^2\mathbf{C}\mathbf{C}' = V(\hat{\boldsymbol{\beta}}) + \sigma^2\mathbf{C}\mathbf{C}',$$
and since $\mathbf{C}\mathbf{C}'$ is positive semidefinite, $V(\tilde{\boldsymbol{\beta}}) - V(\hat{\boldsymbol{\beta}}) \succeq \mathbf{0}$. $\;\square$

*(Note on MSE.)* For a vector estimator the criterion is the mean-squared-error (MSE) **matrix** $E[(\hat{\boldsymbol{\theta}}-\boldsymbol{\theta})(\hat{\boldsymbol{\theta}}-\boldsymbol{\theta})']$ (or its trace); for a scalar it reduces to $E[(\hat{\theta}-\theta)^2] = \text{Var} + \text{Bias}^2$.

### 3. Efficiency
An estimator is **efficient** if it attains the minimum variance among *all* (not only linear) unbiased estimators. If $\boldsymbol{\varepsilon}$ is normal (A4), OLS reaches the **Cramér–Rao lower bound**, so it is efficient.

## 9. Large-Sample (Asymptotic) Properties of OLS

These describe behaviour as the sample size $n \to \infty$.

### 1. Asymptotic unbiasedness
The bias vanishes in the limit:
$$\lim_{n\to\infty} E(\hat{\boldsymbol{\theta}}_n) = \boldsymbol{\theta}.$$
OLS is (exactly) unbiased, hence trivially asymptotically unbiased.

### 2. Consistency
An estimator is **consistent** if it concentrates on the truth as $n$ grows. A sufficient condition is asymptotic unbiasedness **together with** vanishing variance:
$$\lim_{n\to\infty} E(\hat{\boldsymbol{\theta}}_n) = \boldsymbol{\theta} \quad\text{and}\quad \lim_{n\to\infty} Var(\hat{\boldsymbol{\theta}}_n) = \mathbf{0} \;\;\Longrightarrow\;\; \operatorname*{plim}_{n\to\infty} \hat{\boldsymbol{\theta}}_n = \boldsymbol{\theta}.$$
For OLS, $V(\hat{\boldsymbol{\beta}})=\sigma^2(\mathbf{X}'\mathbf{X})^{-1} = \tfrac{\sigma^2}{n}\left(\tfrac{\mathbf{X}'\mathbf{X}}{n}\right)^{-1} \to \mathbf{0}$ provided $\tfrac{1}{n}\mathbf{X}'\mathbf{X}\to\mathbf{Q}$ (a finite, positive-definite matrix). Hence OLS is **consistent** even without normality.

### 3. Asymptotic normality
By a central limit theorem, regardless of the error distribution,
$$\sqrt{n}\,(\hat{\boldsymbol{\beta}} - \boldsymbol{\beta}) \;\xrightarrow{d}\; N\!\left(\mathbf{0},\ \sigma^2 \mathbf{Q}^{-1}\right), \qquad \mathbf{Q} = \operatorname*{plim} \tfrac{1}{n}\mathbf{X}'\mathbf{X}.$$
This justifies using normal/$t$ critical values in large samples *even when A4 does not hold*.

### 4. Asymptotic efficiency
OLS attains the asymptotic Cramér–Rao bound, so it is **asymptotically efficient**.

## 10. Unbiased Estimator of $\sigma^{2}$

### 1. Derivation

The OLS residual vector is
$$\mathbf{e} \equiv \mathbf{y} - \mathbf{X}\hat{\boldsymbol{\beta}} = \mathbf{M}\mathbf{y} = \mathbf{M}\boldsymbol{\varepsilon}, \qquad \mathbf{M} = \mathbf{I} - \mathbf{X}(\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'.$$
(The last equality uses $\mathbf{M}\mathbf{X}=\mathbf{0}$.) Since $\mathbf{M}$ is symmetric and idempotent,
$$E(\mathbf{e}'\mathbf{e}) = E(\boldsymbol{\varepsilon}'\mathbf{M}\boldsymbol{\varepsilon}) = \sigma^2\,\operatorname{tr}(\mathbf{M}) = \sigma^2\,(n-p),$$
because $\operatorname{tr}(\mathbf{M}) = n - \operatorname{tr}(\mathbf{P}) = n - p$. Therefore
$$E\!\left(\frac{\mathbf{e}'\mathbf{e}}{n-p}\right) = \sigma^2 \;\;\Longrightarrow\;\; \boxed{\,s^2 = \frac{\mathbf{e}'\mathbf{e}}{n-p}\,}$$
is **unbiased**. The divisor $n-p$ is the **degrees of freedom**: $p$ are "used up" in estimating $\boldsymbol{\beta}$.

### 2. Property of the ML Estimator $\tilde{\sigma}^{2}$

The ML variance estimator divides by $n$:
$$\tilde{\sigma}^2 = \frac{(\mathbf{y}-\mathbf{X}\tilde{\boldsymbol{\beta}})'(\mathbf{y}-\mathbf{X}\tilde{\boldsymbol{\beta}})}{n} = \frac{\mathbf{e}'\mathbf{e}}{n} = \frac{n-p}{n}\,s^2.$$
Hence $E(\tilde{\sigma}^2) = \tfrac{n-p}{n}\sigma^2 < \sigma^2$: it is **negatively biased** (it underestimates $\sigma^2$). 

The bias factor $\tfrac{n-p}{n}\to 1$, so $\tilde{\sigma}^2$ is **consistent** and the bias is negligible in large samples.

## 11. Confidence Interval for $\boldsymbol{\beta}$

A confidence interval gives a *formal* measure of the precision/reliability of an estimated coefficient.

From $V(\hat{\boldsymbol{\beta}}) = \sigma^2(\mathbf{X}'\mathbf{X})^{-1}$, take the diagonal elements
$$\operatorname{diag}(\mathbf{X}'\mathbf{X})^{-1} = (q_{11}, q_{22}, \ldots, q_{pp}),$$
so that, under normality, $\hat{\beta}_j \sim N(\beta_j,\ \sigma^2 q_{jj})$. 

Because $\sigma$ is unknown we replace it with $s$; standardizing then turns the normal into a **Student's $t$** (the extra uncertainty from estimating $\sigma$ widens the tails):
$$\hat{\beta}_j \sim N(\beta_j, \sigma^2 q_{jj}) \;\longrightarrow\; \frac{\hat{\beta}_j - \beta_j}{\sigma\sqrt{q_{jj}}} \sim N(0,1) \;\longrightarrow\; \frac{\hat{\beta}_j - \beta_j}{s\sqrt{q_{jj}}} \sim t_{n-p}.$$

With probability $(1-\lambda)$,
$$-t_{n-p,\,\lambda/2} \le \frac{\hat{\beta}_j - \beta_j}{s\sqrt{q_{jj}}} \le +t_{n-p,\,\lambda/2},$$
which rearranges into the **confidence interval**
$$\hat{\beta}_j - t_{n-p,\,\lambda/2}\, s\sqrt{q_{jj}} \;\le\; \beta_j \;\le\; \hat{\beta}_j + t_{n-p,\,\lambda/2}\, s\sqrt{q_{jj}}.$$

- $1-\lambda$: **confidence level** — the long-run frequency with which such intervals cover the true $\beta_j$.
- $\lambda$: **significance level** — commonly 5% or 1% (giving 95% or 99% confidence).

*(Interpretation caveat.) The randomness is in the interval, not in $\beta_j$: a 95% interval means 95% of intervals constructed this way contain the fixed true value.*

## 12. Statistical Hypothesis Testing

- The (two-sided) null hypothesis of "no effect" is $H_0:\ \beta_j = 0$, with alternative $H_1:\ \beta_j \neq 0$.
- We know $\dfrac{\hat{\beta}_j - \beta_j}{s\sqrt{q_{jj}}} \sim t_{n-p}$.
- Under $H_0$ this becomes the **$t$-statistic**
$$t_j = \frac{\hat{\beta}_j}{s\sqrt{q_{jj}}} \sim t_{n-p},$$
which follows a $t$-distribution with $(n-p)$ degrees of freedom.
$$t_j = \frac{\hat{\beta}_j}{s\sqrt{q_{jj}}} \sim t_{n-p} \;\longrightarrow\; -t_{n-p,\,\lambda/2} \le t_j \le +t_{n-p,\,\lambda/2}.$$

If the $t$-value is so far from zero that it would occur by chance only, say, 5% of the time, we **reject** $H_0$.

Typical critical values for $\lambda = 0.05$ (5% significance / 95% confidence):
$$t_{n-p,\,\lambda/2} = 1.96 \ (\text{if } n-p=\infty), \qquad = 2.00 \ (\text{if } n-p=60).$$

**Practical notes.**
- **$p$-value**: the probability, under $H_0$, of a $t$-statistic at least as extreme as observed; reject when $p < \lambda$. Equivalently, reject when $0$ lies outside the confidence interval — testing and interval estimation are two views of the same calculation.
- **Joint tests**: to test several coefficients simultaneously (e.g., $H_0:\beta_1=\beta_2=0$) use an **$F$-test**.
- **Goodness of fit**: the coefficient of determination $R^2 = 1 - \dfrac{\mathbf{e}'\mathbf{e}}{\sum_i (y_i-\bar y)^2}$ reports the share of variance explained, but a high $R^2$ does **not** imply a correct or causal model.

## 13. Prediction with Linear Regression

### 1. Settings
Model and assumptions:
$$\mathbf{y} = \mathbf{X}\boldsymbol{\beta} + \boldsymbol{\varepsilon}, \qquad E(\boldsymbol{\varepsilon})=\mathbf{0},\quad E(\boldsymbol{\varepsilon}\boldsymbol{\varepsilon}')=\sigma^2\mathbf{I}.$$

Estimate by OLS/ML: $\hat{\boldsymbol{\beta}} = (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\mathbf{y}$.

For a future/new unit with regressor vector $\mathbf{x}_f$, the **point prediction** is
$$\hat{Y}_f = \mathbf{x}_f'\hat{\boldsymbol{\beta}}.$$

It is useful to distinguish two questions: predicting the *mean* $E[Y_f\mid\mathbf{x}_f]=\mathbf{x}_f'\boldsymbol{\beta}$ (a confidence interval), versus predicting an *individual* outcome $Y_f$ (a wider **prediction** interval that also includes the new error $\varepsilon_f$).

### 2. Strict Model of Future Prediction

$$Y_f = \mathbf{x}_f'\boldsymbol{\beta} + \varepsilon_f$$

- $\mathbf{x}_f$ : independent variable(s) assumed in the future
- $Y_f$ : dependent variable in the future (a random variable)
- $\varepsilon_f$ : future error

The task is to obtain $\hat{Y}_f$, an estimate/predictor of $Y_f$, and to quantify its error.

## 14. Best Linear Unbiased Prediction (BLUP)

A desirable predictor satisfies:

1. **Linearity**: $\hat{Y}_f = \mathbf{a}'\mathbf{y}$.
2. **Unbiasedness**: $E(\hat{Y}_f - Y_f) = 0$.
3. **Best**: it minimizes the mean squared prediction error (equivalently, the variance of the prediction error):
$$\min.\ E\!\left[(\hat{Y}_f - Y_f)^2\right].$$

### 1. Assumption
$$Y_f = \mathbf{x}_f'\boldsymbol{\beta} + \varepsilon_f,$$
in analogy with $E(\boldsymbol{\varepsilon})=\mathbf{0},\ E(\boldsymbol{\varepsilon}\boldsymbol{\varepsilon}')=\sigma^2\mathbf{I}$, so that
$$E(\varepsilon_f) = 0, \qquad E(\varepsilon_f^2) = \sigma^2, \qquad E(\varepsilon_f \boldsymbol{\varepsilon}) = \mathbf{0}.$$

### 2. Result
The BLUP coincides with the plug-in prediction
$$\hat{Y}_f = \mathbf{x}_f'\hat{\boldsymbol{\beta}},$$
and the **variance of the prediction error** is
$$\hat{\sigma}_f^2 = \sigma^2\!\left[\,\underbrace{1}_{\text{new shock }\varepsilon_f} + \underbrace{\mathbf{x}_f'(\mathbf{X}'\mathbf{X})^{-1}\mathbf{x}_f}_{\text{estimation uncertainty in }\hat{\boldsymbol{\beta}}}\right].$$

**Reading the two terms.** 
- Even with perfectly known $\boldsymbol{\beta}$ (second term $=0$) an irreducible error of variance $\sigma^2$ remains because of the future shock $\varepsilon_f$. 
- The second term adds the uncertainty from having estimated $\boldsymbol{\beta}$, and it grows when $\mathbf{x}_f$ lies far from the centre of the observed data — extrapolation is less reliable.

## 16. Worked Example in R

This example ties the theory to computation on a small **fixed** dataset ($n=10$, fully reproducible). It:

(a) estimates the model with `lm()`;
(b) reports the coefficient table, $R^2$, and $F$-statistic via `summary()`;
(c) reports 95% confidence intervals with `confint()`;
(d) reproduces $\hat{\boldsymbol{\beta}} = (\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\mathbf{y}$ by hand;
(e) computes $s^2$ and the standard errors $\sqrt{\operatorname{diag}(s^2(\mathbf{X}'\mathbf{X})^{-1})}$ and checks they match `summary()`;
(f) forms a 95% prediction interval at a new point.

**Running this as R.** This notebook uses the **R (IRkernel)** kernel — select it via *Kernel ▸ Change kernel ▸ R* (Jupyter) or the kernel picker (VS Code). The cell is plain base R (no extra packages). If the R kernel is not yet available, install it once from an R console:

```r
install.packages("IRkernel")
IRkernel::installspec()   # registers the "R" kernel for Jupyter
```

````r
# A reproducible example on a small fixed dataset (n = 10)
x1 <- c(10, 12, 14, 15, 18, 20, 22, 25, 28, 30)
x2 <- c(50, 47, 55, 60, 52, 58, 49, 63, 57, 61)
y  <- c(-4.4, -3.3, -2.8, -4.4, 1.7, 0.0, 5.6, 2.6, 7.6, 7.0)

fit <- lm(y ~ x1 + x2)          # (a) estimation
summary(fit)                    # (b) coefficients, SE, t, p, R^2, F
confint(fit)                    # (c) 95% confidence intervals

X <- cbind(1, x1, x2)           # (d) manual OLS: (X'X)^{-1} X'y
beta_hat <- solve(t(X) %*% X) %*% t(X) %*% y

e  <- y - X %*% beta_hat        # (e) s^2 and standard errors
s2 <- as.numeric(t(e) %*% e) / (length(y) - ncol(X))
sqrt(diag(s2 * solve(t(X) %*% X)))

predict(fit, data.frame(x1 = 20, x2 = 55),    # (f) prediction interval
        interval = "prediction", level = 0.95)
````

## 16. Summary & Connection to the Next Lecture

**Key takeaways.**
- The OLS estimator $\hat{\boldsymbol{\beta}}=(\mathbf{X}'\mathbf{X})^{-1}\mathbf{X}'\mathbf{y}$ is the orthogonal projection of $\mathbf{y}$ onto the regressors.
- Under (A0)–(A3) OLS is **unbiased and BLUE** (Gauss–Markov); adding normality (A4) makes it efficient and gives exact $t$/$F$ inference.
- Asymptotically, OLS is **consistent and asymptotically normal** under weak conditions, so inference is approximately valid in large samples even without normality.
- $s^2=\mathbf{e}'\mathbf{e}/(n-p)$ is the unbiased variance estimator; confidence intervals and $t$-tests follow from $V(\hat{\boldsymbol{\beta}})=\sigma^2(\mathbf{X}'\mathbf{X})^{-1}$.
- Prediction error combines irreducible noise ($\sigma^2$) and estimation uncertainty.

**The crucial limitation → next lecture.** Every result above rests on **exogeneity** (A1): $\mathbf{X}$ uncorrelated with $\boldsymbol{\varepsilon}$. When an omitted variable, reverse causality, or measurement error breaks this, OLS is biased and a coefficient is *not* a causal effect. **Lecture 3 (Basics of Causal Effect Analysis)** formalizes what a causal effect is and how it differs from a regression slope.

**References.** Kmenta, *Elements of Econometrics* (2nd ed.), Ch. 7 (single-equation regression) and Ch. 10 (generalized/large-sample results).
