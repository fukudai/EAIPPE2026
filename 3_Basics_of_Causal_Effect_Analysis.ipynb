- $Y$ : Test score (Outcome [dependent] variable)
- $X$ : Eating breakfast (Policy [independent] variable)
  - $x = 1$ : yes, $x = 0$ : no

Causal Effect for “Observational” Data: 
 $E\left\lbrack Y \right|X = 1,C = c\rbrack - E\left\lbrack Y \right|X = 0,C = c\rbrack$
- $C$ : Covariates or Control variables (e.g., Home environment)

## Simple Linear Regression for Causal Analysis

Simple Linear Regression: 
$E\left\lbrack Y\ \right|X = x\rbrack = \beta_{0} + \beta_{1}x\$
Then, the causal impact is: 
 $E\left\lbrack Y\ \right|X = 1\rbrack - E\left\lbrack Y\ \right|X = 0\rbrack = (\beta_{0} + \beta_{1}) - \beta_{0} =$
This causal impact is an “expected” (average) one.
In reality, the outcome variable may be fluctuated for some reasons: 
 $Y = \beta_{0} + \beta_{1}X + U\text{~~~~~~}\text{or}\text{~~~~~~}Y_{i} = \beta_{0} + \beta_{1}X_{i} + U_{i}\ (i = 1,\ldots,n)$ 
This is called “population model” with the three stochastic variables $(Y,\ X,\ U)$

<!-- slide 5 -->

## Conditions for Causal Detection by Regression

If any of external conditions (e.g., home environment) that affect outcome variable are NOT fully controlled, then $\beta_{1}$ can only exhibits “correlation” relationship and thus we cannot say that “There may be a causal impacts from $X$ to $Y$ ”.
The argument that $\beta_{1}$ is an average causal effect of “Eating Breakfast” on “Test Score” of children is justified if and only if all other external conditions are fully controlled.
This can be formulated with the property of the error term $U$ as:
Assumption 1: $E\lbrack U|X\rbrack = E\lbrack U\rbrack$
In the breakfast example, it means $E\lbrack U|X = 1\rbrack = E\lbrack U|X = 0\rbrack$ although in reality, it might be $E\left\lbrack U \middle| X = 1 \right\rbrack > E\lbrack U|X = 0\rbrack$ with observational data.

<!-- slide 6 -->

## Conditions for Causal Detection by Regression

For making analysis simplified, one more assumption is imposed:
Assumption 2: $E\lbrack U\rbrack = 0$
This assumption is trivial if the intercept term $\beta_{0}$ is included in the regression equation.
- E.g., If $E\lbrack U\rbrack = 1$ , then we make the intercept term $\beta_{0} + 1$ and construct a new error variable $V \equiv U - 1$ leading to: $E\left\lbrack V \right\rbrack = E\left\lbrack U - 1 \right\rbrack = E\left\lbrack U \right\rbrack - 1 = 1 - 1 = 0$ .
By combining Assumptions 1 and 2, we have: $E\lbrack U|X\rbrack = E\lbrack U\rbrack = 0$ 
and it is identical to 
 $E\left\lbrack Y \middle| X \right\rbrack = E\left\lbrack \beta_{0} + \beta_{1}X + U\ \right|\ X\rbrack = \beta_{0} + \beta_{1}X + E\left\lbrack U \middle| X \right\rbrack = \beta_{0} + \beta_{1}X$

## Alternative Method of Estimating Regression Model

Parameters are called identifiable if regression parameters 
(e.g., $\beta_{0},\beta_{1}$ ) can be constructed from the “expectations” of 
the various combinations of stochastic variables $(Y,\ X)$ .
If identifiable, parameters can be estimated by replacing these expectations with their counterpart of sample means computed from observations $\left( Y_{i},\ X_{i} \right)\ i = 1,\ldots,n$
From Assumptions 1 and 2, we have $\text{𝐶𝑜𝑣}\left\lbrack X,U \right\rbrack \equiv$ 
 $E\left\lbrack \text{𝑋𝑈} \right\rbrack - E\lbrack X\ \rbrack E\lbrack U\rbrack = E\lbrack\text{𝑋𝑈}\rbrack = E\lbrack E\lbrack\text{𝑋𝑈}|X\rbrack\rbrack = E\lbrack\text{𝑋𝐸}\lbrack U|X\rbrack\rbrack = 0$ 
that is, $\text{𝐶𝑜𝑣}\lbrack X,U\rbrack = 0$
Furthermore, we have $E\left\lbrack \text{𝑋𝑈} \right\rbrack = 0$ if $E\left\lbrack U \right\rbrack = 0$ .

<!-- slide 9 -->

## Moment Conditions (Method of Moments)

By using $U = Y - \beta_{0} - \beta_{1}X$ , we have the following moment conditions:
$$E\left\lbrack U \right\rbrack = 0\ \Longrightarrow E\lbrack Y - \beta_{0} - \beta_{1}X\rbrack = 0$$
$$E\lbrack\text{𝑋𝑈}\rbrack = 0 \Longrightarrow E\lbrack X(Y - \beta_{0} - \beta_{1}X)\rbrack = 0$$
These are the simultaneous equations with respect to $\beta_{0},\beta_{1}$ and we solve it as:
$$E\lbrack Y - \beta_{0} - \beta_{1}X\rbrack\ = 0 \Longrightarrow E\lbrack Y\rbrack - \beta_{0} - \beta_{1}E\lbrack X\rbrack = 0 \Longrightarrow$$
$$E\left\lbrack X\left( Y - \beta_{0} - \beta_{1}X \right) \right\rbrack = 0 \Longrightarrow E\left\lbrack \text{𝑋𝑌} \right\rbrack - \beta_{0}E\left\lbrack X \right\rbrack - \beta_{1}E\left\lbrack \text{𝑋𝑋} \right\rbrack = \cdots$$
$$=$$
Consequently, we have:
- $\beta_{1} = \frac{\text{𝐶𝑜𝑣}\left\lbrack X,Y \right\rbrack}{V\left\lbrack X \right\rbrack}$ (Assuming $V\left\lbrack X \right\rbrack \neq 0$ ) and $\beta_{0} = E\lbrack Y\rbrack - \frac{\text{𝐶𝑜𝑣}\left\lbrack X,Y \right\rbrack}{V\left\lbrack X \right\rbrack}E\lbrack X\rbrack$
- If we replace these expectations with their counterpart sample mean, they are equivalent to the ones estimated with ordinary least square method.

<!-- slide 10 -->

## Equivalence to OLS

For the dataset $\left( Y_{i},X_{i} \right)\ i = 1,\ldots n$ , we estimate parameters by OLS:
$$\beta_{1} = \frac{\text{𝐶𝑜𝑣}\left\lbrack X,Y \right\rbrack}{V\left\lbrack X \right\rbrack}\ \Longrightarrow \ {\widehat{\beta}}_{1} = \frac{\frac{1}{n}\sum_{i = 1}^{n}{\left( x_{i} - \overline{x} \right)\left( y_{i} - \overline{y} \right)}}{\frac{1}{n}\sum_{i = 1}^{n}\left( x_{i} - \overline{x} \right)^{2}\ }$$
$$\beta_{0} = E\left\lbrack Y \right\rbrack - \frac{\text{𝐶𝑜𝑣}\left\lbrack X,Y \right\rbrack}{V\left\lbrack X \right\rbrack}E\left\lbrack X \right\rbrack\ \Longrightarrow \ {\widehat{\beta}}_{0} = \overline{y} - {\widehat{\beta}}_{1}\overline{x}$$
As noted at the last class, the OLS estimator has some good properties for identifying the exact (true) causal impacts such as:
unbiasedness, consistency, efficiency and asymptotic normality.
For experimental data, the simple regression model (e.g., Y=b0+b1X) can estimate causal effects without any biases.

## Multiple Linear Regression

Assume that the observational data includes external factors (e.g., detailed home environment information) for investigating the impact of breakfast eating on test score.
- The questionnaire asks, for example, parents’ educational level, household income, learning condition at home, etc.
- We call them “Covariates” instead of explanatory variables
Then we can approach to the estimation of genuine causal impacts
$Y$ : Outcome variable, $X$ : Policy variable, and $C$ : Covariate
Actually, both $X$ and $C$ are explanatory variables, i.e., $\left( X,\ C \right) = (X_{1},\ X_{2})$

<!-- slide 13 -->

## Multiple Linear Regression

Conditional expectation of outcome: $E\lbrack Y|X,\ C\rbrack$
We model this as: $E\left\lbrack Y \middle| X,\ C \right\rbrack = \beta_{0} + \beta_{1}X + \beta_{2}C$
This is rewritten as a population model: $Y = \beta_{0} + \beta_{1}X + \beta_{2}C + U$
In general, more explanatory variables can be added (multiple linear regression).
- $C$ is, for example, household income.
Now it is possible to investigate how much the test score raises:
- By eating breakfast while the household income is controlled ( $\beta_{1}$ ), or
- By increasing the household income in one unit while eating breakfast is controlled ( $\beta_{2}$ ).
“Ceteris Paribus” (in Latin) approach = All other things being equal.

<!-- slide 14 -->

## Estimating Multiple Linear Regression Model

Assume there are $k$ explanatory variables (= $k + 1$ parameters):
 $Y = \beta_{0} + \beta_{1}X_{1} + \beta_{2}X_{2} + \cdots + \beta_{k}X_{k} + U$
We then have $k + 1$ moment conditions: 
 $E\left\lbrack U \middle| X_{1} \right\rbrack = E\left\lbrack U \middle| X_{2} \right\rbrack = \cdots = E\left\lbrack U \middle| X_{k} \right\rbrack = E\left\lbrack U \right\rbrack = 0$
The solving procedure is the same with the case of simple regression
The evaluation of goodness-of-fit by the coefficient of determination:
 $R^{2} = 1 - \frac{\sum_{i = 1}^{n}{\widehat{u}}_{i}^{2}}{\sum_{i = 1}^{n}\left( y_{i} - \overline{y} \right)^{2}}$ where $\ {\widehat{u}}_{i}$ is the residual
By considering the degree of freedom ( $n - k - 1$ ), it is adjusted:
 $R_{a}^{2} = 1 - \frac{\left( 1 - R^{2} \right)\left( n - 1 \right)}{n - k - 1}$

## Bias due to Omitted Variables

True population model: $Y = \beta_{0} + \beta_{1}X + \beta_{2}C + U$
But in reality, it is very likely that covariates are NOT fully known for various reasons. Then, $C$ is omitted (“omitted variables”).
Consequently, we can only estimate the model: $Y = \beta_{0} + \beta_{1}X + V$ 
but where $V = \beta_{2}C + U$
Simple linear regression result: ${\widehat{\beta}}_{1} = \frac{\frac{1}{n}\sum_{i = 1}^{n}{\left( x_{i} - \overline{x} \right)\left( y_{i} - \overline{y} \right)}}{\frac{1}{n}\sum_{i = 1}^{n}\left( x_{i} - \overline{x} \right)^{2}\ }$
By substituting the true model, however, we obtain: 
 $E\left\lbrack {\widehat{\beta}}_{1} \right\rbrack = \beta_{1} + \beta_{2}\gamma_{1} \neq \beta_{1}$ 
( $\gamma_{1}$ : regression coefficient for $C = \gamma_{0} + \gamma_{1}X + U_{c}$ ). Thus ${\widehat{\beta}}_{1}$ is biased
Bias Term
