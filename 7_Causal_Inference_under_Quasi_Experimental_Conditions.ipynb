## Utilizing Institutions and/or Customs to find Control Group

One of the primary condition for evaluating the effect of public policies is to control external conditions appropriately, that is, to find appropriate control groups.
On the contrary, idealized experiments are hard to be conducted in many cases.
When we carefully look at various public policies or relevant institutions or customs, 
regarding it as “quasi-experimental conditions”, we sometimes see that:
- Their external conditions are almost the same (very similar).
- For some reasons (e.g., rules, institutions or customs), some people are affected by the policy of interests while other people are unaffected.
If there is a “point of discontinuity” which separates almost similar individuals into control group and treatment group, we can apply “Regression Discontinuity Design” (RDD) to appropriate estimate the causal impacts of the policy.

## Empirical Analysis Using Educational System in Israel (Angrist & Lavy 1999)

Class size (i.e., number of the students per class) tends to be larger in urban area than in rural area.
But the effects of urban characteristics (e.g., higher accessibility to educational opportunities other than schooling, income level differences of parents) on scholastic achievement are not fully considered.
Schools with higher reputation attracts more students in the neighborhood. If so, better schools may have more students, leading to the increase in their class size.
This is a typical “spurious correlation” between test score and class size without appropriately controlled by “Good School” conditions.

## Maimonides’ Rule

It is highly possible that (1) a school with 40 students and (2) a school with 41 students are almost similar meaning that external conditions are well controlled.
The only difference between these two schools is the difference in class size.
That is a kind of “Quasi” experimental condition coincidentally created by existing rules.

## Maimonides’ Rule and RDD

In their research, Angrist and his co-author used the class-size function induced by Maimonides’ rule to construct instrumental variable estimates of class-size effects.
Although the class-size function and the instruments derived from it are themselves a function of the size of enrollment cohorts, these functions are nonlinear and nonmonotonic.
They could therefore control for a wide range of enrollment effects when using the rule as an instrument.
The resulting evidence for a causal impact of class size on test scores is strengthened by the fact that even when controlling for other enrollment effects, the up-and-down pattern in the class size/ enrollment size relationship induced by Maimonides' rule matches a similar pattern in test scores.
Since it seems unlikely that enrollment effects other than those working through class size would generate such a pattern, Maimonides' rule provides an unusually credible source of exogenous variation for class-size research.
This sort of identification argument can be viewed as an application of RDD for evaluation of class size effect on test scores.
In other words, this is a kind of “Quasi”-experimental condition coincidentally created by existing rules.
Source: Angrist and Lavy (1999)

## Sharp RDD or Fuzzy RDD?

RDD utilizes the discontinuous change in the policy variable due to the relevant rule and focuses on the nearby sample observations around such discontinuous point called “threshold” (e.g., 40 students per class) to estimate the causal impact.
Sharp RDD
- Certainly the policy variables are changed discontinuously at that threshold.
- Simply compare the outcome means between below-group and above-group of the threshold
- Applied to the “strict” discontinuity cases due to, for example, laws
Fuzzy RDD
- Actually, the example of Angrist and Lavy (1999) is a Fuzzy RDD.
- Probabilistically the policy variables are changed discontinuously at that threshold.
- The causal impacts can be estimated as the mean outcome difference divided by the probability differences between before and after of threshold.
- Possible to estimate by regression analysis
- Applied to the “soft” discontinuity cases (e.g., customs or guidelines)

$C_{i}$ is the number of students for the school year of student $i$ ,

We can construct the linear regression:
$$Y_{i} = \beta_{0} + \beta_{1}X_{i} + \beta_{2}C_{i} + U_{i}$$
Then $\beta_{1}$ is interpreted at the causal effect of “Introducing small-sized class” on scholastic achievement.
For the case of Fuzzy, we can perform two-stage least square (2SLS) methods for the same regression formula by using the instrumental variables that affect on the probability that $X_{i} = 1$ .

## Summary of Regression Discontinuity Design

If we focus on the “point of discontinuity” (threshold) in the value of an outcome variable due to rules or customs, it is sometimes possible to regard that (1) the before the threshold is a control group and 
(2) the after the threshold is a treatment group while external conditions of these groups are almost similar.
We may try to find such “Jumping Point” due to rules, making the environment to be a quasi-experimental condition.
How can RDD be used for infrastructure policy evaluation?

Example: Regulation of Automobile Fuel Economy 
 in Japan (Ito & Sallee 2018)
- “The fuel economy” (possible mileage of driving per unit gasoline consumption,

- The fuel economy regulation required for a given vehicle is a decreasing step function of its weight in Japanese law.

## Bunching Analysis: Fuel-Economy Example

Japanese fuel-economy standards, which were first introduced in 1979 and have changed four times since, are vehicle weight-based. (c.f., Footprint [vehicle-area size] based in the USA)
The data used in Ito and Sallee (2018), which begin in 2001 in Japan, span the two most recent policy regimes.
To be in compliance with the regulation, automobile manufacturing firms must have a sales-weighted average fuel economy that exceeds the sales-weighted average target of their vehicles, given their weights. Firms are required to meet this standard only in the “target year” of a policy.

## Bunching Analysis: Fuel-Economy Example

Automakers therefore have a large incentive to increase vehicle weight only up to key thresholds where the mandated fuel economy drops discretely.
These notches do not change the fundamental economic incentives at play, but they aid empirical identification.
Our first empirical strategy is to test for “bunching” (excess mass) in the distribution of vehicle weight around regulatory thresholds
Qualitatively, this implies that vehicle weight is responsive to policy incentives, which possibly implies significant economic loss.

## Bunching Analysis: Fuel-Economy Example

There is a strong evidence of weight manipulation by firms in response to the policy because there is visible excess mass at the notch points.
Between Panel A and B, the mass points shifted precisely in accordance with the change in the locations of the notch points. This shift in the location of the distributional anomalies provides further compelling evidence that firms respond to the attributed-based regulation.
Ito and Sallee (2018) estimate that 10% of Japanese vehicles have had their weight increased in response to the policy.
Among the affected vehicles, they estimate that weight rose by 110 kilograms on average by the fuel-regulation policy.
