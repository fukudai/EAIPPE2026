## What is the “Panel Data”?

The type of dataset we have analyzed so far is called “Cross-Sectional data”.
- One-shot observations for the population.
Panel Data is the dataset like that:
- Annual income trajectories 2009-2024 for Graduates of UT Civil Eng.
- Daily passenger counts at stations along Nanboku Line during a year.
- GDP Changes for 1980-2020 in OECD countries, etc.
Panel data is thus the Repeated Cross-Sectional Data for the Same Observation Units.
But when we simply say, “Repeated Cross-Sectional Data”, it is not necessarily the data from the same but different observation units.

## Panel Data as a Repeated Cross-Sectional Data

“Data Pooling”: Constructing the larger dataset by combining some datasets. For example,
- Pooling the data of Tokyo Metropolitan and the data of Kanagawa Prefecture
- Pooling the data of 1990 data and 2000 data of Tokyo Metropolitan 
(This is called Pooled Cross-Sectional Data)
Advantage of Panel data (1): possible to increase the sample size by pooling several datasets. [(2) appears later]
In the pooled CS data, the data from the same individual but in different years are regarded as the observation from different individual.

## Panel Data as a Repeated Cross-Sectional Data

Example:
- We want to know if the average income of Tokyo residents has changed between 1990 and 2000 or not.
- Pooling the 1990 dataset and the 2000 dataset.
- Then perform a multiple regression for the pooled dataset with the following specifications:
- $Y_{i}\ = \beta_{0} + \beta_{1}x_{1i} + \cdots + \beta_{Y}\text{𝐷𝑈𝑀𝑀}Y_{2000,i} + \epsilon_{i}$ 
 $\text{where}\ \text{𝐷𝑈𝑀𝑀}Y_{2000,i}$ is 1 if the observation $i$ is in 2000, and 0 in 1990.
- Then the change in the average income of $Y_{i}$ between 1990 and 2000 can be estimated as ${\widehat{\beta}}_{Y}$ with the standard OLS.

## Difference-in-Differences (DID) Estimator

Example:
- We want to know if the construction of garbage (waste) disposal sites would lower the land price in the neighborhood or not.
- How would we investigate the causal impacts of the construction on the land price?
- It is not sufficient to simply compare the land price nearby the sites and the land price distant from the sites. Why?
- Straightforward Intuition: In what kind of places did garbage disposal sites tend to be constructed?
- If that intuition is true, the land price of the construction site might have been already lower before it was constructed.

## DID Estimator

Possible Answer:
- Compare “the temporal change” in the land price between the nearby place and the distant place from the garbage disposal construction site.
- $Y$ : Outcome variable (e.g., land price)
- $\delta_{\text{𝐷𝐷}}$ : Causal effect measured with DID Estimator
$$\delta_{\text{𝐷𝐷}} = \left( Y_{\text{𝐷𝑖𝑠𝑡𝑎𝑛𝑡},\ \text{𝐴𝑓𝑡𝑒𝑟}} - Y_{\text{𝐷𝑖𝑠𝑡𝑎𝑛𝑡},\ \text{𝐵𝑒𝑓𝑜𝑟𝑒}} \right) - \left( Y_{\text{𝑁𝑒𝑎𝑟𝑏𝑦},\ \text{𝐴𝑓𝑡𝑒𝑟}} - Y_{\text{𝑁𝑒𝑎𝑟𝑏𝑦},\ \text{𝐵𝑒𝑓𝑜𝑟𝑒}} \right)$$
$$= \left( 280 - 320 \right) - \left( 200 - 290 \right)$$
$$= - 40 - ( - 90) = 50$$
$$\delta_{\text{𝐷𝐷}} = \left( Y_{\text{𝐷𝑖𝑠𝑡𝑎𝑛𝑡},\ \text{𝐴𝑓𝑡𝑒𝑟}} - Y_{\text{𝑁𝑒𝑎𝑟𝑏𝑦},\ \text{𝐴𝑓𝑡𝑒𝑟}} \right) - \left( Y_{\text{𝐷𝑖𝑠𝑡𝑎𝑛𝑡},\ \text{𝐵𝑒𝑓𝑜𝑟𝑒}} - Y_{\text{𝑁𝑒𝑎𝑟𝑏𝑦},\ \ \text{𝐵𝑒𝑓𝑜𝑟𝑒}} \right)$$
$$= \left( 280 - 200 \right) - \left( 320 - 290 \right)$$
$$= 80 - 30 = 50$$

## Testing with DID

Treatment Group is the group affected by the policy (e.g., Nearby).
Control Group is the group not affected by the policy (e.g., Distant).
If each observation point of land price is randomly assigned to either Treatment group or Control group, then the policy effect (i.e., the impact of the garbage disposal site on the land price) is estimated simply by comparing the temporal changes in the average outcome values between Treatment group and Control group.
DID Estimator: For time periods 1, 2 and for groups A, B, we compute
$$\left( Y_{2,B} - Y_{2,A} \right) - \left( Y_{1,B} - Y_{1,A} \right)\text{~~~~}\left\lbrack \text{or}\ \left( Y_{2,B} - Y_{1,B} \right) - \left( Y_{2,A} - Y_{1,A} \right) \right\rbrack$$
Possible to infer only with four average numbers

## Validity of the Random Assignment Assumption

In some cases, individuals are NOT randomly assigned to each group, but INTENTIONALLY belong to a particular group.
This is called “Self-Selection”.
Example: Analyzing the effect of Job-Training Program on Earnings
- In many cases, it is each individual’s decision-making whether he/she will participate in the program or not.
- Consequently, those who participate in the program are likely to be the person with the higher willingness or ability (i.e., larger labor productivity).
- If it is the case, the higher wage or income would be observed for those who participated in the program, even though the program has actually little effect.
The DID with the regression can reduce that bias to some extent.

<!-- slide 15 -->

## Self-Selection Problems

There is no problem if it is possible to control all external factors that are correlated with both program participation and outcome variable.
But it is often that unobservable factors are correlated with program participation decision (e.g., individual’s productivity).
If so, the program participation is BIASED and thus the DID estimator cannot fully capture the true causal impact of the policy.
C.f., If all individuals are randomly assigned to either control group or treatment group, it is called “Natural Experiment”.

## Two-Periods Panel Data

Advantage of Panel data (2): Possible to perform the regression analysis that accommodates the effect of the unobservable (time-invariant) factors.
This may lead to excluding the omitted variable bias.
Model: $Y_{\text{𝑖𝑡}} = \beta_{0} + \beta_{1}x_{\text{𝑖𝑡}1} + \cdots + \beta_{k}x_{\text{𝑖𝑡𝑘}} + a_{i} + u_{\text{𝑖𝑡}}$
$ai$ is unobservable and individual ( $i$ ) specific factor, 
called individual effect. It is also called an unobserved fixed effect.
In the standard OLS regression, we assume $\upsilon\text{𝑖𝑡} = a_{i} + u_{\text{𝑖𝑡}}$ as error term (instead of $u_{\text{𝑖𝑡}}$ ) though it is biased when $ai$ is correlated with explanatory variables.

<!-- slide 17 -->

## First-Differences Approach

Now we take difference between two time-periods: 
 $\Delta Yi = \beta 1\Delta xi1 + \cdots + \beta k\Delta x\text{𝑖𝑘} + \Delta ui$
where $\Delta Yi = Y_{i2} - Y_{i1},\text{~~}\Delta xij = x_{i2j} - x_{i1j}\ \forall j,\text{~~}\Delta ui = u_{i2} - u_{i1}$
Now we see that the unobservable $ai\$ is cancelled out!!
If $\Delta xij$ and $\Delta ui$ are uncorrelated, the standard OLS regression can obtain the unbiased estimator of $\beta$ s.
Note that the above regression formula does not contain the constant term ( $\beta_{0}$ ). But it is possible to incorporate “time-specific constant” as an extra term: $``\delta_{0} \times \text{𝐷𝑈𝑀𝑀}Y_{t = 2}"$ where $\text{𝐷𝑈𝑀𝑀}Y_{t = 2}$ take 1 if that observation is from time-period 2 and zero from time-period 1. 
This is called “time-specific fixed effect”.

<!-- slide 18 -->

## What if more than three time-periods?

Model: $Y_{\text{𝑖𝑡}} = \beta_{0} + \beta_{1}x_{1\text{𝑖𝑡}} + \ldots + \beta_{k}x_{\text{𝑘𝑖𝑡}} + a_{i} + u_{\text{𝑖𝑡}}\$
Possible to use panel data sets for the many (i.e., $T\left( > 1 \right)$ ) periods
For each $i$ , we compute average over $T$ periods:
$$Y_{i} = \beta_{0} + \beta_{1}x_{1i} + \ldots + \beta_{k}x_{\text{𝑘𝑖}} + a_{i} + u_{i}\$$
where $Y_{i} \equiv \frac{1}{T}\sum_{t = 1}^{T}Y_{\text{𝑖𝑡}}$ , $x_{\text{𝑘𝑖}} \equiv \frac{1}{T}\sum_{t = 1}^{T}x_{\text{𝑘𝑖𝑡}}$ , $u_{i} \equiv \frac{1}{T}\sum_{t = 1}^{T}u_{\text{𝑖𝑡}}$
Take the difference between each period (t) and the average:
$$(Y_{\text{𝑖𝑡}} - Y_{i}) = \beta_{1}(x_{1\text{𝑖𝑡}} - x_{1i}) + \ldots + \beta_{k}(x_{\text{𝑘𝑖𝑡}} - x_{\text{𝑘𝑖}})\ + (u_{\text{𝑖𝑡}} - u_{i})$$
If all $u_{\text{𝑖𝑡}}$ s are uncorrelated with all explanatory variables (strong exogeneity), the standard OLS can achieve the consistent estimators.

## Conclusions and future works

Conclusions
- Employment in the retail and wholesale industry increased from 2.4 to 8.2 percent in municipality connected by the high-speed rail network.
- It has been shown that a one percent increase in accessibility due to high-speed rail would lead to 0.1 percent growth in retail and wholesale employment.
- We find that one percent increase accessibility due to highway would lead from 0.6 to 1.2 percent growth in manufacture employment.
- This would be because high-speed rail contributes to the transport passenger flow while the highway mainly contributes to logistic flow.
- These result imply that the affected region due to high-speed rail development might be possibly specialized into some specific industry sectors such as retail and wholesale in the long run.
Future works
- We have checked the parallel trend assumption following Lin(2017) and Dong(2018). However, it is difficult to check the parallel trend assumption exactly.
    - →To check the parallel trend assumption exactly, we have to use more appropriate empirical strategies such as synthetic control method.

## R 実習コード（Class5_DID.R）

対応するRスクリプトをIRkernelで実行できます．データの相対パスは適宜調整して下さい．

```r
# Adapted from "Differences-in-Differences (using R)" by Oscar Torres-Reyna
# https://www.princeton.edu/~otorres/DID101R.pdf

# Data Read
Panel101 <- read.table("Panel101.csv", header = TRUE, sep= ",")

# Overview of the data and summary/descriptive statistics
colnames(Panel101)
table(Panel101$country)
table(Panel101$year)
min(Panel101$year)
max(Panel101$year)
summary(Panel101$y)
mean(Panel101$y)
median(Panel101$y)
sd(Panel101$y) #standard deviation
install.packages("plotrix")
library(plotrix)
std.error(Panel101$y) #standard error
summary(Panel101$x1)
summary(Panel101$x2)
summary(Panel101$x3)

# Create a dummy variable to indicate the years after the treatment.
# Lets assume that treatment started in 1994. 
# In this case, years before 1994 will have a value of 0 and 1994+ a 1. 
Panel101$time <- ifelse(Panel101$year >= 1994, 1, 0)

# Create a dummy variable to identify the group exposed to the treatment. 
# Lets assumed that countries with code E,F, and G were treated (=1). 
# The other countries were not treated (=0).
Panel101$treated<- ifelse(Panel101$country == "E" | 
                            Panel101$country == "F" | 
                            Panel101$country == "G", 1, 0)

# Create an interaction between time and treated. 
# We will call this interaction 'did'.
Panel101$did <- Panel101$time * Panel101$treated

# Estimating the DID estimator
didreg <- lm(y ~ treated + time + did, data = Panel101)
summary(didreg)

# The coefficient for 'did' is the DID estimator. 
# The effect is significant at 10% with the treatment having a negative effect.

# Estimating the DID estimator 
# (using the multiplication method) 
# The output is the same
didreg1 <- lm(y ~ treated*time, data = Panel101) 
summary(didreg1)

# The coefficient for 'treated#time' is the DID estimator ('did' in the previous example). 
# The effect is significant at 10% with the treatment having a negative effect.

# Adding the covariates to the model
didreg2 <- lm(y ~ treated + time + treated*time + x1 + x2 + x3, data = Panel101) 
summary(didreg2)
```
